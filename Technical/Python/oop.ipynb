## 1. OOP Principles

The four pillars of Object-Oriented Programming:

- **Encapsulation** – bundling data and methods together, hiding internal state.
- **Abstraction** – exposing only what's necessary, hiding complexity.
- **Inheritance** – a class derives behavior from a parent class.
- **Polymorphism** – different classes respond to the same interface differently.

---

## 2. Class Syntax

```python
class Dog:
    species = "Canis familiaris"      # class attribute

    def __init__(self, name, age):
        self.name = name              # instance attribute
        self.age = age

    def bark(self):
        return f"{self.name} says Woof!"

d = Dog("Rex", 3)
print(d.bark())   # Rex says Woof!
```

---

## 3. Inheritance Syntax

A child class inherits attributes and methods from a parent.

```python
class Animal:
    def speak(self):
        return "..."

class Cat(Animal):          # Cat inherits Animal
    def speak(self):
        return "Meow"

class Kitten(Cat):
    pass

k = Kitten()
print(k.speak())  # Meow  (inherited from Cat)
```

Use `super()` to call the parent's implementation:

```python
class Dog(Animal):
    def speak(self):
        base = super().speak()
        return f"{base} + Woof"
```

## 4. Class vs Instance

| Concept | Scope | Shared? |
|---|---|---|
| Class attribute | Defined on the class | Yes, all instances |
| Instance attribute | Defined via `self` | No, per object |

```python
class Counter:
    count = 0                    # class attribute

    def __init__(self):
        Counter.count += 1
        self.id = Counter.count  # instance attribute

a, b = Counter(), Counter()
print(Counter.count)  # 2
print(a.id, b.id)     # 1  2
```

## 5. Operator Overloading

Redefine how operators work for your objects by implementing dunder methods.

```python
class Vector:
    def __init__(self, x, y):
        self.x, self.y = x, y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __repr__(self):
        return f"Vector({self.x}, {self.y})"

print(Vector(1, 2) + Vector(3, 4))  # Vector(4, 6)
```

## 6. Callable Objects

An object becomes callable by defining `__call__`.

```python
class Multiplier:
    def __init__(self, factor):
        self.factor = factor

    def __call__(self, value):
        return value * self.factor

double = Multiplier(2)
print(double(5))   # 10
print(callable(double))  # True
```

## 7. Dunder (Magic) Notation

Methods wrapped in double underscores (`__method__`) are called **dunder** (double underscore) or **magic** methods. Python calls them automatically in specific situations.

```python
len([1, 2, 3])      # calls list.__len__()
a + b               # calls a.__add__(b)
str(obj)            # calls obj.__str__()
obj()               # calls obj.__call__()
```

## 8. Class Attributes vs Instance Attributes

```python
class Config:
    debug = False         # class-level, shared

    def __init__(self, name):
        self.name = name  # instance-level, unique

c1 = Config("app1")
c2 = Config("app2")

Config.debug = True       # affects all instances
print(c1.debug)  # True
print(c2.debug)  # True

c1.debug = False          # creates a NEW instance attribute, shadows class attr
print(c1.debug)  # False  (instance shadows class)
print(c2.debug)  # True   (still reads class attr)
```

## 9. `@property` Decorator

Allows methods to be accessed like attributes. Enables controlled getters and setters.

```python
class Circle:
    def __init__(self, radius):
        self._radius = radius

    @property
    def radius(self):
        return self._radius

    @radius.setter
    def radius(self, value):
        if value < 0:
            raise ValueError("Radius must be positive")
        self._radius = value

    @property
    def area(self):
        return 3.14159 * self._radius ** 2

c = Circle(5)
print(c.area)    # 78.53...
c.radius = 10
```

## 10. DRY — Don't Repeat Yourself

Avoid code duplication. Extract shared logic into a single place (method, base class, mixin).

```python
# BAD
class Dog:
    def describe(self):
        print(f"Name: {self.name}, Age: {self.age}")

class Cat:
    def describe(self):
        print(f"Name: {self.name}, Age: {self.age}")

# GOOD — single source of truth
class Animal:
    def describe(self):
        print(f"Name: {self.name}, Age: {self.age}")

class Dog(Animal): pass
class Cat(Animal): pass
```

## 11. Standard Method Decorators

### `@classmethod`

Receives the **class** (`cls`) as the first argument instead of the instance. Used for alternative constructors or class-level operations.

```python
class Date:
    def __init__(self, y, m, d):
        self.y, self.m, self.d = y, m, d

    @classmethod
    def from_string(cls, s):        # alternative constructor
        y, m, d = map(int, s.split("-"))
        return cls(y, m, d)

dt = Date.from_string("2024-01-15")
```

### `@staticmethod`

No implicit first argument. A utility function logically grouped in the class but not needing `self` or `cls`.

```python
class MathUtils:
    @staticmethod
    def is_prime(n):
        return n > 1 and all(n % i for i in range(2, int(n**0.5)+1))

print(MathUtils.is_prime(7))  # True
```

## 12. Attribute Hiding (`_` and `__`)

| Convention | Meaning |
|---|---|
| `_name` | Protected — "internal use", not enforced |
| `__name` | Private — name-mangled to `_ClassName__name` |

```python
class BankAccount:
    def __init__(self, balance):
        self._balance = balance     # protected (convention)
        self.__pin = 1234           # private (name-mangled)

acc = BankAccount(100)
print(acc._balance)           # 100 (accessible but discouraged)
# print(acc.__pin)            # AttributeError
print(acc._BankAccount__pin)  # 1234 (mangled name still accessible)
```

## 13. Abstract Base Classes (ABC)

Force subclasses to implement specific methods. Using `abc` module.

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self) -> float: ...

class Rectangle(Shape):
    def __init__(self, w, h):
        self.w, self.h = w, h

    def area(self):
        return self.w * self.h

# Shape()        # TypeError: can't instantiate abstract class
print(Rectangle(3, 4).area())  # 12
```

## 14. Attribute Access (`getattr`, `setattr`)

Dynamic attribute access at runtime.

```python
class Config:
    host = "localhost"
    port = 8080

c = Config()
print(getattr(c, "host"))               # localhost
print(getattr(c, "missing", "default")) # default (fallback)
setattr(c, "port", 9090)
print(c.port)                           # 9090
```

Also relevant: `hasattr(obj, name)`, `delattr(obj, name)`.

## 15. Common Dunder Methods

### `__init__`, `__new__`, `__del__`

| Method | When called | Purpose |
|---|---|---|
| `__new__` | Before `__init__` | Creates the instance (rarely overridden) |
| `__init__` | After `__new__` | Initializes the instance |
| `__del__` | On garbage collection | Cleanup (avoid relying on it) |

```python
class Singleton:
    _instance = None

    def __new__(cls):
        if cls._instance is None:
            cls._instance = super().__new__(cls)
        return cls._instance
```

---

### `__repr__` and `__str__`

| Method | Used by | Purpose |
|---|---|---|
| `__repr__` | `repr()`, debugger | Unambiguous developer representation |
| `__str__` | `print()`, `str()` | Human-readable string |

```python
class Point:
    def __init__(self, x, y):
        self.x, self.y = x, y

    def __repr__(self):
        return f"Point({self.x!r}, {self.y!r})"

    def __str__(self):
        return f"({self.x}, {self.y})"

p = Point(1, 2)
print(repr(p))  # Point(1, 2)
print(p)        # (1, 2)
```

> Rule of thumb: always implement `__repr__`; implement `__str__` when you need a different human-facing format.

### Comparison Dunders

`__lt__`, `__le__`, `__gt__`, `__ge__`, `__eq__`, `__ne__`

Used to support `<`, `<=`, `>`, `>=`, `==`, `!=`. Enables sorting and comparison logic.

```python
from functools import total_ordering

@total_ordering   # auto-fills missing comparisons from __eq__ + one other
class Temperature:
    def __init__(self, degrees):
        self.degrees = degrees

    def __eq__(self, other):
        return self.degrees == other.degrees

    def __lt__(self, other):
        return self.degrees < other.degrees

temps = [Temperature(30), Temperature(20), Temperature(25)]
print(min(temps).degrees)   # 20
```

---

### `__hash__`

Objects that define `__eq__` should also define `__hash__` to be usable in sets/dicts. If `__eq__` is defined and `__hash__` is not, Python sets `__hash__ = None` (unhashable).

```python
class Point:
    def __init__(self, x, y):
        self.x, self.y = x, y

    def __eq__(self, other):
        return (self.x, self.y) == (other.x, other.y)

    def __hash__(self):
        return hash((self.x, self.y))

s = {Point(1, 2), Point(1, 2)}
print(len(s))  # 1
```

### `__class__`

Returns the class of an object.

```python
x = 42
print(x.__class__)           # <class 'int'>
print(x.__class__.__name__)  # int
```

### `__bool__`

Controls truthiness. If absent, Python falls back to `__len__`.

```python
class Wallet:
    def __init__(self, balance):
        self.balance = balance

    def __bool__(self):
        return self.balance > 0

w = Wallet(0)
print(bool(w))   # False
if not w:
    print("Wallet is empty")
```

### Arithmetic Operators

`__add__`, `__sub__`, `__mul__`, `__truediv__`, `__floordiv__`, `__pow__`, `__mod__`

Also reflected variants: `__radd__`, `__rsub__`, etc. (called when the left operand doesn't support the operation).

```python
class Money:
    def __init__(self, amount):
        self.amount = amount

    def __add__(self, other):
        return Money(self.amount + other.amount)

    def __mul__(self, factor):
        return Money(self.amount * factor)

    def __rmul__(self, factor):   # supports: 3 * Money(10)
        return self.__mul__(factor)

    def __repr__(self):
        return f"${self.amount}"

print(Money(10) + Money(5))  # $15
print(3 * Money(10))         # $30
```

## 16. `__slots__`

Restricts instance attributes to a fixed set. Saves memory and speeds up attribute access by avoiding the per-instance `__dict__`.

```python
class Point:
    __slots__ = ("x", "y")   # no __dict__ created

    def __init__(self, x, y):
        self.x, self.y = x, y

p = Point(1, 2)
# p.z = 3  # AttributeError: 'Point' has no attribute 'z'
```

Use when creating many instances (e.g., game entities, data records) and memory matters.

## 17. MRO — Method Resolution Order

Defines the order Python searches for methods in an inheritance hierarchy. Inspect it via `ClassName.__mro__` or `ClassName.mro()`.

### Diamond Problem

```
      A
     / \
    B   C
     \ /
      D
```

Without MRO rules, calling `D.method()` is ambiguous — should it use `B` or `C`?

### C3 Linearization Algorithm

Python uses **C3** to compute a consistent, predictable order:

> Order = class itself + merge of parents' MROs + list of parents, using a left-to-right, no-reversal rule.

```python
class A:
    def hello(self): print("A")

class B(A):
    def hello(self): print("B")

class C(A):
    def hello(self): print("C")

class D(B, C): pass

D().hello()       # B  (B comes before C in D's definition)
print(D.__mro__)  # D -> B -> C -> A -> object
```

## 18. Mixins

Small, focused classes that add a specific capability and are not meant to be instantiated alone. Combined via multiple inheritance.

```python
class JsonMixin:
    def to_json(self):
        import json
        return json.dumps(self.__dict__)

class LogMixin:
    def log(self):
        print(f"[LOG] {self.__class__.__name__}: {self.__dict__}")

class User(JsonMixin, LogMixin):
    def __init__(self, name, age):
        self.name, self.age = name, age

u = User("Alice", 30)
print(u.to_json())   # {"name": "Alice", "age": 30}
u.log()              # [LOG] User: {'name': 'Alice', 'age': 30}
```

## 19. Composition vs Inheritance

| | Inheritance | Composition |
|---|---|---|
| Relationship | "is-a" | "has-a" |
| Coupling | Tight | Loose |
| Reuse | Via subclassing | Via delegation |
| Flexibility | Less | More |

```python
# Inheritance — Dog IS-A Animal
class Animal:
    def breathe(self): print("breathing")

class Dog(Animal): pass

# Composition — Car HAS-A Engine
class Engine:
    def start(self): print("vroom")

class Car:
    def __init__(self):
        self.engine = Engine()   # composed, not inherited

    def start(self):
        self.engine.start()
```

> Prefer composition over inheritance when the relationship is "has-a" or behavior needs to change at runtime.

## 20. SOLID Principles

| Letter | Principle | Short definition |
|---|---|---|
| **S** | Single Responsibility | A class should have only one reason to change |
| **O** | Open/Closed | Open for extension, closed for modification |
| **L** | Liskov Substitution | Subclasses must be substitutable for their base class |
| **I** | Interface Segregation | Prefer small, specific interfaces over large ones |
| **D** | Dependency Inversion | Depend on abstractions, not concretions |
```python
# S — Single Responsibility example
# Violation: one class handles both user data and email
class User:
    def save(self): ...
    def send_welcome_email(self): ...  # not its responsibility!

# Fix: split into two classes
class User:
    def save(self): ...

class UserMailer:
    def send_welcome_email(self, user): ...
```
```python
# O — Open/Closed example
# Violation: adding a new shape requires modifying existing code
class AreaCalculator:
    def calculate(self, shape):
        if isinstance(shape, Circle): ...
        elif isinstance(shape, Square): ...  # must edit this every time!

# Fix: extend via polymorphism
class Shape:
    def area(self): ...

class Circle(Shape):
    def area(self): return 3.14 * self.radius ** 2

class Square(Shape):
    def area(self): return self.side ** 2
```
```python
# L — Liskov example (violation)
class Bird:
    def fly(self): ...
class Penguin(Bird):
    def fly(self):
        raise NotImplementedError  # breaks substitution!
# Fix: separate the interface
class FlyingBird(Bird): ...
class Penguin(Bird): ...  # doesn't inherit fly
```
```python
# I — Interface Segregation example
# Violation: one fat interface forces unused methods
class Worker:
    def work(self): ...
    def eat(self): ...  # robots don't eat!

class Robot(Worker):
    def eat(self): pass  # forced to implement irrelevant method

# Fix: split into focused interfaces
class Workable:
    def work(self): ...

class Eatable:
    def eat(self): ...

class Human(Workable, Eatable): ...
class Robot(Workable): ...
```
```python
# D — Dependency Inversion example
# Violation: high-level module depends on a concrete class
class MySQLDatabase:
    def save(self, data): ...

class UserService:
    def __init__(self):
        self.db = MySQLDatabase()  # tightly coupled!

# Fix: depend on an abstraction
class Database:
    def save(self, data): ...

class UserService:
    def __init__(self, db: Database):
        self.db = db  # any Database implementation works
```

## 21. Composition, Aggregation, Association

These describe **relationships between objects**, not classes.

| Relationship | Lifetime dependency | Example |
|---|---|---|
| **Association** | Independent | Teacher ↔ Student |
| **Aggregation** | Child can exist without parent | Team has Players (players exist independently) |
| **Composition** | Child cannot exist without parent | House has Rooms (rooms don't exist outside the house) |

```python
# Association — loose reference
class Driver:
    def __init__(self, car):
        self.car = car  # just uses it

# Aggregation — owns but doesn't destroy
class Team:
    def __init__(self, players):
        self.players = players  # players created outside

# Composition — creates and owns
class House:
    def __init__(self):
        self.rooms = [Room() for _ in range(5)]  # rooms owned by house
```
