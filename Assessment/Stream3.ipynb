## Section 1: Time Complexity & Big O

**Q1.** What is the time complexity of `x = n * 2` and why?

<details><summary>Answer</summary>

**O(1) — Constant time.** The operation runs once regardless of the value of `n`.

</details>

---

**Q2.** What is the time complexity of a single `for i in range(n)` loop?

<details><summary>Answer</summary>

**O(n) — Linear time.** The loop body executes exactly `n` times.

</details>

---

**Q3.** Does changing the step in `for i in range(0, n, 5)` change the Big O complexity?

<details><summary>Answer</summary>

**No.** The loop runs `n/5` times, but the constant factor `1/5` is dropped. It is still **O(n)**.

</details>

---

**Q4.** What is the time complexity of this loop, and why?

```python
while n > 0:
    n = n // 2
```

<details><summary>Answer</summary>

**O(log n) — Logarithmic time.** `n` is halved on each iteration, so the number of iterations is approximately log₂(n).

</details>

---

**Q5.** What is the time complexity of two nested `for` loops both running `n` times?

<details><summary>Answer</summary>

**O(n²) — Quadratic time.** The inner loop executes `n` times for each of the `n` outer iterations: `n × n = n²`.

</details>

---

**Q6.** Rank from fastest to slowest: O(n²), O(1), O(n log n), O(log n), O(n).

<details><summary>Answer</summary>

**O(1) < O(log n) < O(n) < O(n log n) < O(n²)**

</details>

## Section 2: Software Engineering Practices

**Q7.** What is a code review and what are its main benefits?

<details><summary>Answer</summary>

A systematic peer examination of source code. **Benefits:** catches bugs early, enforces coding standards, and spreads knowledge across the team.

</details>

---

**Q8.** What is the difference between Static and Dynamic Code Analysis?

<details><summary>Answer</summary>

- **Static:** analyzes code *without executing it* (e.g., ESLint, SonarQube). Finds common bugs early, but can have false positives.
- **Dynamic:** analyzes code *during execution* (e.g., Valgrind). Finds real runtime issues like memory leaks and race conditions.

</details>

---

**Q9.** What is the difference between Continuous Integration, Continuous Delivery, and Continuous Deployment?

<details><summary>Answer</summary>

- **CI:** Every merge triggers automated build + tests.
- **Continuous Delivery:** Every passing build is *automatically deployable* (but manually released).
- **Continuous Deployment:** Every passing build is *automatically released* to production.

</details>

---

**Q10.** What are three common bottlenecks in a CI/CD pipeline and how do you address them?

<details><summary>Answer</summary>

1. **Long test suites** → parallelize or split into tiers.
2. **Heavy Docker builds** → use layer caching.
3. **Sequential stages that could run together** → use DAG-style pipelines.

</details>

---

**Q11.** What is the difference between a centralized VCS (e.g., SVN) and a distributed VCS (e.g., Git)?

<details><summary>Answer</summary>

- **Centralized:** single server; limited offline work; network-dependent.
- **Distributed:** every clone is a full repo; supports full offline operations; lightweight branching.

</details>

---

**Q12.** What is the difference between `git merge` and `git rebase`?

<details><summary>Answer</summary>

- **merge:** creates a merge commit combining two branch histories.
- **rebase:** replays commits on top of another branch, producing a linear history.

</details>

---

**Q13.** Describe the GitFlow branching strategy.

<details><summary>Answer</summary>

Uses long-lived branches: `main` (production), `develop`, `feature/*`, `release/*`, and `hotfix/*`. Features merge into `develop`; releases merge into both `main` and `develop`. Best for scheduled releases but can be complex.

</details>

---

**Q14.** What is Trunk-Based Development and when should it be used?

<details><summary>Answer</summary>

Everyone commits to `main` (or very short-lived branches). Incomplete work is hidden behind feature flags. Best for high-velocity teams with strong CI/CD.

</details>

---

**Q15.** Describe the Testing Pyramid and the ideal proportion of each test type.

<details><summary>Answer</summary>

- **Unit tests (base):** many, fast, cheap — test a single function in isolation.
- **Integration tests (middle):** fewer, medium cost — test components working together.
- **E2E tests (top):** very few, slow, expensive — simulate real user flows. Most tests should be unit tests.

</details>

---

**Q16.** What is the difference between a unit test and an integration test?

<details><summary>Answer</summary>

- **Unit test:** isolates a single function/class; dependencies are mocked.
- **Integration test:** verifies that multiple components work correctly together (e.g., service + real database).

</details>

---

**Q17.** Name the five types of performance testing and briefly describe each.

<details><summary>Answer</summary>

1. **Load testing:** simulate expected load.
2. **Stress testing:** push beyond limits to find the breaking point.
3. **Spike testing:** sudden surge of users.
4. **Soak/Endurance testing:** sustained load over a long period (detects memory leaks).
5. **Scalability testing:** gradually increase load to observe scaling behavior.

</details>

---

**Q18.** What is the Definition of Done (DoD) and why is it important?

<details><summary>Answer</summary>

A shared checklist of criteria a task must meet before it can be called complete (e.g., code reviewed, tests passing, deployed to staging, docs updated). It prevents ambiguity and enforces quality gates.

</details>

---

**Q19.** What is the Definition of Ready (DoR)?

<details><summary>Answer</summary>

A checklist ensuring a task is ready to be worked on — for example: story written, acceptance criteria defined, designs available. It prevents teams from pulling incomplete work into a sprint.

</details>

## Section 3: Software Design

**Q20.** Name four architectural patterns and describe when you would use each.

<details><summary>Answer</summary>

- **Layered:** clear separation (Presentation → Business → Data); good for enterprise apps.
- **Microservices:** independent deployable services per domain; good for scalable cloud-native apps.
- **Event-Driven:** async communication via events; good for real-time, decoupled systems.
- **MVC:** Model-View-Controller; good for UI-heavy apps.

</details>

---

**Q21.** What are the SOLID principles? Give a one-line definition of each.

<details><summary>Answer</summary>

- **S — Single Responsibility:** one class, one reason to change.
- **O — Open/Closed:** open for extension, closed for modification.
- **L — Liskov Substitution:** subtypes must be replaceable for their base type.
- **I — Interface Segregation:** many small interfaces > one general interface.
- **D — Dependency Inversion:** depend on abstractions, not concretions.

</details>

---

**Q22.** What does DRY stand for and what problem does it solve?

<details><summary>Answer</summary>

**Don't Repeat Yourself.** It avoids code duplication by keeping a single source of truth for logic, reducing maintenance cost and the risk of bugs in multiple copies.

</details>

---

**Q23.** What is YAGNI and when does it apply?

<details><summary>Answer</summary>

**You Aren't Gonna Need It.** Don't build functionality for hypothetical future requirements. Apply it when you feel tempted to add flexibility or abstraction that nobody has asked for yet.

</details>

---

**Q24.** What is Dependency Injection and why does it improve testability?

<details><summary>Answer</summary>

DI means dependencies are provided from outside (injected) rather than created internally. This decouples classes and allows test code to inject mock/stub dependencies, making units independently testable.

</details>

---

**Q25.** What is the Circuit Breaker pattern and why is it useful in distributed systems?

<details><summary>Answer</summary>

It stops cascading failures: after a service fails repeatedly, the circuit "opens" and fast-fails all calls for a period, instead of waiting and consuming resources. This protects the rest of the system.

</details>

---

**Q26.** What is the Saga pattern?

<details><summary>Answer</summary>

A pattern for managing distributed transactions across microservices. Instead of a single ACID transaction, a saga is a sequence of local transactions, each publishing events or messages to trigger the next step, with compensating actions if a step fails.

</details>

---

**Q27.** What is the difference between composition and inheritance? When should you prefer composition?

<details><summary>Answer</summary>

- **Inheritance ("is-a"):** child inherits from parent — tight coupling.
- **Composition ("has-a"):** class holds a reference to another — loose coupling.

Prefer composition when the relationship is "has-a", when behavior needs to change at runtime, or to avoid deep, brittle inheritance hierarchies.

</details>

---

**Q28.** What is the Observer design pattern?

<details><summary>Answer</summary>

A behavioral pattern where an object (subject) notifies a list of dependents (observers) automatically when its state changes. Used for event listeners, pub/sub systems, and MVC data binding.

</details>

---

**Q29.** What is the Strategy design pattern?

<details><summary>Answer</summary>

A behavioral pattern that defines a family of algorithms, encapsulates each one, and makes them interchangeable at runtime. Example: `sorter.setStrategy(new QuickSort())`.

</details>

---

**Q30.** What are the three basic design principles: KISS, DRY, and YAGNI?

<details><summary>Answer</summary>

- **KISS:** Keep It Simple — avoid unnecessary complexity.
- **DRY:** Don't Repeat Yourself — single source of truth.
- **YAGNI:** Don't build for hypothetical future requirements.

</details>

**Q51.** What is the CAP Theorem?

<details><summary>Answer</summary>

In a distributed system, you can only guarantee **two of three**: **Consistency** (every read returns the latest write), **Availability** (every request gets a response), and **Partition Tolerance** (system works despite network splits). Since partitions are unavoidable, the real choice is CP vs AP.

</details>

</details>

## Section 6: Python OOP

**Q53.** What are the four pillars of OOP?

<details><summary>Answer</summary>

**Encapsulation** (hide internal state), **Abstraction** (expose only what's necessary), **Inheritance** (derive behavior from a parent), **Polymorphism** (different classes respond to the same interface differently).

</details>

---

**Q54.** What is the difference between a class attribute and an instance attribute?

<details><summary>Answer</summary>

- **Class attribute:** defined on the class itself; shared by all instances.
- **Instance attribute:** defined via `self` in `__init__`; unique per object.

If you assign to an instance with the same name as a class attribute, Python creates a new instance attribute that **shadows** the class one.

</details>

---

**Q55.** What does `super()` do in Python?

<details><summary>Answer</summary>

It returns a proxy to the **parent class**, allowing you to call its methods. Commonly used in `__init__` to run the parent's initialization before adding child-specific logic.

</details>

---

**Q56.** What is Method Resolution Order (MRO) in Python and which algorithm determines it?

<details><summary>Answer</summary>

MRO defines the order Python searches for methods in an inheritance hierarchy. It is determined by the **C3 Linearization** algorithm. Inspect it with `ClassName.__mro__`. It resolves the diamond problem consistently.

</details>

---

**Q57.** What are dunder (magic) methods? Give three examples.

<details><summary>Answer</summary>

Methods wrapped in double underscores that Python calls automatically in specific situations. Examples: `__init__` (construction), `__str__` (string representation), `__add__` (operator `+`).

</details>

---

**Q58.** What is the difference between `__repr__` and `__str__`?

<details><summary>Answer</summary>

- `__repr__`: unambiguous developer-facing representation; used by `repr()` and in the debugger.
- `__str__`: human-readable; used by `print()` and `str()`.

Rule: always implement `__repr__`; add `__str__` only when a different user-facing format is needed.

</details>

---

**Q59.** What is the `@property` decorator used for?

<details><summary>Answer</summary>

It allows a method to be accessed like an attribute. Combined with `@<name>.setter`, it enables **controlled getters and setters** with validation, without changing the calling code.

</details>

---

**Q60.** What is the difference between `@classmethod` and `@staticmethod`?

<details><summary>Answer</summary>

- `@classmethod`: receives `cls` as the first argument; can access and modify class state. Used for alternative constructors.
- `@staticmethod`: no implicit first argument; a utility function grouped in the class but not needing `self` or `cls`.

</details>

---

**Q61.** What is the difference between `_name` and `__name` in Python?

<details><summary>Answer</summary>

- `_name`: **protected** by convention — signals "internal use," not enforced.
- `__name`: **private** — Python name-mangles it to `_ClassName__name`, preventing direct external access.

</details>

---

**Q62.** What is an Abstract Base Class (ABC) and how do you create one?

<details><summary>Answer</summary>

An ABC forces subclasses to implement specific methods. Use the `abc` module: inherit from `ABC` and decorate abstract methods with `@abstractmethod`. Instantiating the ABC directly raises a `TypeError`.

</details>

---

**Q63.** What is `__slots__` and when would you use it?

<details><summary>Answer</summary>

`__slots__` restricts instance attributes to a fixed set, preventing the creation of `__dict__` per instance. Use it when creating many instances (e.g., game entities, data records) where memory efficiency matters.

</details>

---

**Q64.** What is a Mixin and how does it differ from regular inheritance?

<details><summary>Answer</summary>

A Mixin is a small, focused class that adds a specific capability (e.g., `to_json()`) and is not meant to be instantiated alone. Combined via multiple inheritance. Unlike regular parent classes, mixins have no "is-a" relationship with the classes that use them.

</details>

---

**Q65.** Explain the difference between association, aggregation, and composition.

<details><summary>Answer</summary>

- **Association:** objects reference each other with no lifetime dependency (Teacher ↔ Student).
- **Aggregation:** parent "owns" child but child can exist independently (Team has Players).
- **Composition:** parent creates and fully owns child; child cannot exist without parent (House has Rooms).

</details>

---

**Q66.** What happens to `__hash__` when you define `__eq__` in a class?

<details><summary>Answer</summary>

Python sets `__hash__ = None`, making instances **unhashable** (cannot be used in sets or as dict keys). You must define `__hash__` explicitly if you need the object to be hashable.

</details>

---

**Q67.** What is `__bool__` used for? What does Python fall back to if it's not defined?

<details><summary>Answer</summary>

`__bool__` controls the truthiness of an object (`bool(obj)`). If not defined, Python falls back to `__len__`: object is truthy if `len() != 0`.

</details>

## Section 7: Python Functions

**Q68.** What is the LEGB rule in Python?

<details><summary>Answer</summary>

The order Python uses to resolve variable names: **L**ocal → **E**nclosing → **G**lobal → **B**uilt-in.

</details>

---

**Q69.** What is the difference between `global` and `nonlocal` keywords?

<details><summary>Answer</summary>

- `global`: declares that a variable refers to the module-level (global) name.
- `nonlocal`: declares that a variable refers to the nearest enclosing (non-global) scope, enabling a nested function to modify it.

</details>

---

**Q70.** What is a closure in Python? Give an example.

<details><summary>Answer</summary>

A closure is an inner function that **remembers variables from its enclosing scope** even after the outer function has returned.

```python
def multiplier(factor):
    def multiply(x):
        return x * factor  # 'factor' is captured
    return multiply

double = multiplier(2)
double(5)  # 10
```

</details>

---

**Q71.** What are `*args` and `**kwargs`?

<details><summary>Answer</summary>

- `*args`: collects extra **positional** arguments into a **tuple**.
- `**kwargs`: collects extra **keyword** arguments into a **dict**.

</details>

---

**Q72.** What is a higher-order function? Give two built-in examples.

<details><summary>Answer</summary>

A function that takes a function as an argument or returns one. Built-in examples: `map()`, `filter()`, `sorted(key=...)`.

</details>

---

**Q73.** What does `map(func, iterable)` do? How would you write it as a list comprehension?

<details><summary>Answer</summary>

Applies `func` to every element and returns a lazy iterator.
```python
list(map(lambda x: x * 2, [1, 2, 3]))  # [2, 4, 6]
# Equivalent:
[x * 2 for x in [1, 2, 3]]
```

</details>

---

**Q74.** What does `functools.reduce` do and what is its signature?

<details><summary>Answer</summary>

`reduce(func, iterable)` applies `func` cumulatively to reduce the iterable to a single value. Example: `reduce(lambda a, b: a + b, [1,2,3,4])` → `10`.

</details>

---

**Q75.** What are the pros and cons of recursion in Python?

<details><summary>Answer</summary>

**Pros:** elegant for tree/graph traversal, mirrors mathematical definitions, clean divide-and-conquer code.  
**Cons:** risk of `RecursionError` (default limit ~1000), higher memory usage, slower than iteration (no tail call optimization in Python).

</details>

---

**Q76.** Can you mutate the dict returned by `locals()`? What about `globals()`?

<details><summary>Answer</summary>

- `globals()` — **yes**, mutations affect the actual global namespace.
- `locals()` — **no** (in functions); it returns a snapshot copy. Modifying it does not change local variables.

</details>

---

**Q77.** Why is the exception variable deleted after an `except` block in Python 3?

<details><summary>Answer</summary>

To avoid **reference cycles**. The exception object holds a reference to the traceback, which holds references to the local variables. Python deletes the name automatically after the `except` block. Save it explicitly if needed.

</details>

## Section 8: Python Exceptions

**Q78.** What is the difference between `Exception` and `BaseException`?

<details><summary>Answer</summary>

`BaseException` is the root of all exceptions, including `SystemExit`, `KeyboardInterrupt`, and `GeneratorExit`. `Exception` is the base for all "normal" exceptions. Always catch `Exception` in application code, not `BaseException`, to avoid silencing exits and interrupts.

</details>

---

**Q79.** What does the `else` clause in a `try-except-else` block do?

<details><summary>Answer</summary>

The `else` block runs **only if no exception was raised** in the `try` block. It is used for code that should execute on success but is not part of the guarded operation.

</details>

---

**Q80.** What does the `finally` clause guarantee?

<details><summary>Answer</summary>

It **always executes**, regardless of whether an exception was raised, caught, or a `return` statement was hit. Used for cleanup (closing files, releasing locks).

</details>

---

**Q81.** How do you create a custom exception in Python?

<details><summary>Answer</summary>

Subclass `Exception` (never `BaseException`) and optionally override `__init__` to add context.

```python
class InsufficientFundsError(Exception):
    def __init__(self, amount, balance):
        super().__init__(f"Cannot withdraw {amount}. Balance: {balance}.")
```

</details>

---

**Q82.** What is exception chaining? What is the difference between `raise X from Y` and `raise X from None`?

<details><summary>Answer</summary>

- `raise X from Y`: **explicit chaining** — X is raised as a consequence of Y; both appear in the traceback.
- `raise X from None`: **suppresses chaining** — hides the original exception from the traceback.

</details>

---

**Q83.** Should you use `assert` for production input validation? Why?

<details><summary>Answer</summary>

**No.** `assert` statements are disabled when Python runs with the `-O` (optimize) flag. Use explicit `if/raise` for production validation. Use `assert` only for internal invariants in development/debug.

</details>

## Section 9: Python Testing

**Q84.** What is the difference between a Mock, a Stub, and a Patch?

<details><summary>Answer</summary>

- **Mock:** replaces an object and **records calls** — used to verify interactions.
- **Stub:** returns **predefined data** — used to control indirect inputs.
- **Patch:** **temporarily replaces** something in scope — used to replace external calls (HTTP, DB).

</details>

---

**Q85.** What is TDD (Test-Driven Development) and what are its three phases?

<details><summary>Answer</summary>

Write a **failing test** (Red) → write **minimum code** to pass it (Green) → **Refactor** while keeping tests green.

</details>

---

**Q86.** What are the main advantages of `pytest` over `unittest`?

<details><summary>Answer</summary>

pytest: simpler function-based syntax, no boilerplate, powerful `@pytest.fixture`, `@pytest.mark.parametrize`, rich plugin ecosystem (`pytest-cov`, `pytest-xdist`, `pytest-mock`), excellent output readability.

</details>

---

**Q87.** What causes flaky tests and how do you fix them?

<details><summary>Answer</summary>

**Causes:** time dependencies, shared state, race conditions, real network calls.  
**Fixes:** freeze time (`freezegun`), isolate test state, mock external calls, use deterministic data.

</details>

---

**Q88.** What does test coverage measure, and why is 100% coverage not always a good goal?

<details><summary>Answer</summary>

Coverage measures which lines/branches were executed during tests. 100% coverage does not guarantee correctness — you can have full coverage with meaningless assertions. Use it to identify **untested areas**, not as a proof of quality. Common targets: 80–90% for business logic.

</details>

---

**Q89.** What is contract testing (e.g., Pact) and why is it useful for microservices?

<details><summary>Answer</summary>

Contract testing verifies that a consumer's expected API interactions are compatible with what the provider actually delivers. It catches breaking API changes without requiring both services to be running simultaneously.

</details>

**Q90.** Why should you avoid mocking the database in integration tests?

<details><summary>Answer</summary>

Because mocked DB tests can pass while real migrations or schema changes silently break the integration. Use a real test database (via Docker) to catch actual incompatibilities.

</details>

## Section 10: Python Async & Parallel Programming

**Q91.** What is the difference between asynchronous and parallel execution in Python?

<details><summary>Answer</summary>

- **Async:** single thread; tasks interleave by yielding during I/O waits (cooperative).
- **Parallel:** multiple processes/threads run *simultaneously* on different CPU cores.

Async is best for I/O-bound; parallel (multiprocessing) is best for CPU-bound.

</details>

---

**Q92.** What is the Python GIL and what are its consequences?

<details><summary>Answer</summary>

The **Global Interpreter Lock** is a mutex in CPython that allows only one thread to execute Python bytecode at a time. Consequence: CPU-bound threads don't achieve true parallelism. Solution for CPU-bound work: use `multiprocessing` (each process has its own GIL).

</details>

---

**Q93.** What is the difference between a coroutine, a Task, and a Future in asyncio?

<details><summary>Answer</summary>

- **Coroutine:** an `async def` function; doesn't run until awaited or scheduled.
- **Task:** wraps a coroutine and schedules it to run **concurrently** on the event loop.
- **Future:** a low-level object representing a result not yet computed. Tasks are a subclass of Future.

</details>

---

**Q94.** What does `asyncio.gather()` do?

<details><summary>Answer</summary>

Runs multiple coroutines **concurrently** and waits for all to complete, returning their results as a list. Three tasks with delays of 1s, 2s, 1s finish in ~2s total, not 4s.

</details>

---

**Q95.** When should you use `ThreadPoolExecutor` vs `ProcessPoolExecutor` vs `asyncio`?

<details><summary>Answer</summary>

- `asyncio` → I/O-bound with async-compatible libraries.
- `ThreadPoolExecutor` → I/O-bound blocking code.
- `ProcessPoolExecutor` → CPU-bound computation (bypasses GIL with separate processes).

</details>

---

**Q96.** What is a race condition? Show a simple Python example and its fix.

<details><summary>Answer</summary>

A race condition occurs when the result depends on the unpredictable order of thread execution.

```python
counter = 0
def increment():
    global counter
    counter += 1  # read-modify-write is NOT atomic

# Fix: use a Lock
lock = threading.Lock()
def increment():
    global counter
    with lock:
        counter += 1
```

</details>

---

**Q97.** What is the difference between a `Lock` and a `Semaphore` in threading?

<details><summary>Answer</summary>

- **Lock:** allows only **one** thread into a critical section at a time.
- **Semaphore:** allows up to **N** threads concurrently. Useful for limiting, e.g., max database connections.

</details>

---

**Q98.** What is the difference between cooperative and preemptive multitasking?

<details><summary>Answer</summary>

- **Preemptive (threads):** the OS interrupts a running thread at any moment.
- **Cooperative (asyncio):** a task runs until it *voluntarily* yields at an `await`. One blocking call stalls all tasks.

</details>

---

**Q99.** What is `contextvars.ContextVar` and why is it preferred over `threading.local()` in async code?

<details><summary>Answer</summary>

`ContextVar` provides **per-task isolated state** in asyncio. `threading.local()` is per-thread, but asyncio tasks share a thread, so all tasks would read the same `threading.local()` value. `ContextVar` gives each asyncio Task its own copy automatically.

</details>

---

**Q100.** What is `uvloop` and when should you use it?

<details><summary>Answer</summary>

`uvloop` is a drop-in replacement for asyncio's default event loop, built on libuv (Node.js engine). It can be **2–4x faster**. Use it in high-throughput Linux/macOS production servers (APIs, WebSockets). Not available on Windows.

</details>

## Section 11: Databases

**Q101.** What are the four ACID properties of a transaction?

<details><summary>Answer</summary>

- **Atomicity:** all or nothing — if any step fails, the whole transaction rolls back.
- **Consistency:** DB moves from one valid state to another, respecting all constraints.
- **Isolation:** concurrent transactions don't see each other's intermediate states.
- **Durability:** once committed, data survives crashes.

</details>

---

**Q102.** What are the four SQL transaction isolation levels and the anomaly each prevents?

<details><summary>Answer</summary>

| Level | Dirty Read | Non-Repeatable Read | Phantom Read |
|---|---|---|---|
| READ UNCOMMITTED | Possible | Possible | Possible |
| READ COMMITTED | Prevented | Possible | Possible |
| REPEATABLE READ | Prevented | Prevented | Possible |
| SERIALIZABLE | Prevented | Prevented | Prevented |

</details>

---

**Q103.** What is the difference between `INNER JOIN`, `LEFT JOIN`, and `FULL OUTER JOIN`?

<details><summary>Answer</summary>

- **INNER JOIN:** only rows with matches in **both** tables.
- **LEFT JOIN:** all rows from the left + matching from right (NULL where no match).
- **FULL OUTER JOIN:** all rows from both; NULL where no match on either side.

</details>

---

**Q104.** What is a window function and how does it differ from `GROUP BY`?

<details><summary>Answer</summary>

A window function (e.g., `RANK()`, `AVG() OVER(...)`) performs calculations across related rows **without collapsing them** into one row per group. `GROUP BY` reduces rows; window functions preserve the original row count.

</details>

---

**Q105.** What is normalization? Describe 1NF, 2NF, and 3NF.

<details><summary>Answer</summary>

- **1NF:** all values are atomic; each row is unique.
- **2NF:** 1NF + no partial dependency on a composite primary key.
- **3NF:** 2NF + no transitive dependency (non-key column depending on another non-key column).

</details>

---

**Q106.** What are the trade-offs between normalization and denormalization?

<details><summary>Answer</summary>

- **Normalized:** less storage, faster writes, high integrity — good for **OLTP**.
- **Denormalized:** more storage, faster reads (no JOINs), risk of anomalies — good for **OLAP / reporting**.

</details>

---

**Q107.** What is a database index and what are its trade-offs?

<details><summary>Answer</summary>

An index (usually a B-tree) speeds up data retrieval by avoiding full table scans. **Trade-off:** speeds up `SELECT` but slows down `INSERT`, `UPDATE`, `DELETE` because the index must be maintained on every write.

</details>

---

**Q108.** What is the difference between a Stored Procedure and a Function in SQL?

<details><summary>Answer</summary>

- **Stored Procedure:** can have side effects (INSERT, UPDATE); called with `CALL`/`EXEC`; does not need to return a value.
- **Function:** must return a value; ideally no side effects; can be used in `SELECT` expressions.

</details>

**Q110.** What is the difference between symmetric and asymmetric (public-key) encryption?

<details><summary>Answer</summary>

- **Symmetric:** same key for encrypt and decrypt (e.g., AES). Fast; good for bulk data.
- **Asymmetric:** a key pair — public key encrypts, private key decrypts (e.g., RSA). Used in TLS handshakes to securely exchange a session key.

</details>

**Q111.** What is a deadlock and how do you prevent it?

<details><summary>Answer</summary>

A deadlock occurs when two threads each wait for a lock the other holds. Prevention: always **acquire locks in a consistent order** across all threads.

</details>

---

**Q112.** What is the difference between a Type 1 and Type 2 hypervisor?

<details><summary>Answer</summary>

- **Type 1 (Bare-metal):** runs directly on hardware (e.g., VMware ESXi, KVM). Lower overhead, used in production.
- **Type 2 (Hosted):** runs on a host OS (e.g., VirtualBox, VMware Workstation). Easier setup, used for development.

</details>

**Q113.** What is the difference between a container and a virtual machine?

<details><summary>Answer</summary>

- **VM:** full OS with its own kernel; GBs in size; minutes to start; strong isolation.
- **Container:** shares host OS kernel; MBs in size; seconds to start; process-level isolation.

</details>

**Q120.** What is the difference between `asyncio.create_task()` and `await coroutine()`?
