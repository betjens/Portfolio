# Assessment — Software Engineering & Python


## Part 1 — Software Engineering Practices

**Q1.** What is the difference between static code analysis and dynamic code analysis?
**A:** Static analysis examines source code *without executing it* (e.g., ESLint, Pylint) — it catches syntax errors, style violations, and common bugs at build time. Dynamic analysis examines code *during execution* (e.g., Valgrind, JaCoCo) — it finds real runtime issues like memory leaks and race conditions, but requires the app to actually run.

---

**Q2.** What are three benefits of code reviews?
**A:** (1) Catches defects early before they reach production. (2) Enforces team coding standards and consistency. (3) Spreads knowledge across the team, reducing silos.

---

**Q3.** What is the difference between Continuous Delivery and Continuous Deployment?
**A:** Continuous Delivery means every passing build is *automatically deployable* to staging/prod, but a human approves the release. Continuous Deployment goes further — every passing build is *automatically released* to production with no manual step.

---

**Q4.** Name three common CI/CD pipeline bottlenecks and how to address them.
**A:** (1) Long test suites — parallelize or split into tiers. (2) Heavy Docker builds — use layer caching. (3) Sequential stages that could be independent — use DAG-style pipelines to run them in parallel.

---

**Q5.** What is the key difference between a centralized VCS (e.g., SVN) and a distributed VCS (e.g., Git)?
**A:** In a centralized VCS, there is a single server and clients must be connected for most operations. In a distributed VCS, every clone is a full repository, enabling full offline work (commits, branching, log) without needing network access.

---

**Q6.** What does `git rebase` do and how does it differ from `git merge`?
**A:** `git rebase` replays your commits on top of another branch, producing a linear history with no merge commit. `git merge` integrates two branches with a merge commit that preserves both histories. Rebase gives cleaner history; merge preserves the true order of work.

---

**Q7.** What is `git stash` used for?
**A:** It temporarily saves work-in-progress changes (both staged and unstaged) so you can switch branches or pull updates without committing incomplete work. `git stash pop` restores the saved state.

---

**Q8.** Describe the GitFlow branching strategy. What is it best suited for?
**A:** GitFlow uses long-lived branches: `main` (production), `develop` (integration), `feature/*` (new work), `release/*` (release prep), and `hotfix/*` (urgent fixes). It's best suited for products with scheduled release cycles that need clear release management. The drawback is complexity and potentially painful merges.

---

**Q9.** What is Trunk-Based Development and what does it require to work well?
**A:** Everyone commits to `main` (or very short-lived feature branches). Incomplete features are hidden behind feature flags. It requires strong CI/CD, high test coverage, and team discipline, but enables high-velocity continuous delivery.

---

**Q10.** Explain the Testing Pyramid. What proportion of tests should each level represent?
**A:** The pyramid has three tiers — Unit (base, most numerous: fast, cheap, isolated), Integration (middle, fewer: test component interaction), E2E (top, fewest: slow, expensive, full-stack). Most tests should be unit tests; very few should be E2E. Inverting the pyramid leads to slow, brittle pipelines.

---

**Q11.** What is the difference between a unit test and an integration test?
**A:** A unit test validates a single function or class in isolation — dependencies are mocked. An integration test verifies that multiple components (e.g., a service and a real database) work correctly together. Integration tests are slower but catch wiring bugs that unit tests miss.

---

**Q12.** What is the difference between load testing and stress testing?
**A:** Load testing simulates *expected* user load to validate normal behaviour. Stress testing pushes the system *beyond* normal limits to find its breaking point.

---

**Q13.** Name four types of performance testing beyond load and stress.
**A:** (1) Spike testing — sudden surge of users. (2) Soak/endurance testing — sustained load over a long period to detect memory leaks. (3) Scalability testing — gradually increase load to measure how the system scales. (4) Volume testing — large amounts of data to test storage and processing capacity.

---

**Q14.** What is a "Definition of Done" and why is it important?
**A:** A DoD is a team-agreed checklist of conditions a task must satisfy to be called complete (e.g., code reviewed, tests passing, deployed to staging, docs updated). It prevents ambiguity, ensures consistent quality, and stops half-finished work from being counted as done.

---

**Q15.** What is the purpose of the `fail fast` principle in CI/CD?
**A:** To catch issues as early as possible in the pipeline, reducing the cost and time to fix them. For example, running linting before tests so trivial errors are caught immediately rather than after a long test suite.

---

## Part 2 — Cloud Fundamentals

**Q16.** What is the Shared Responsibility Model in cloud computing?
**A:** Security duties are split between provider and customer. The provider manages physical infrastructure, hypervisors, and networking hardware. The customer manages data, access control, and (depending on service model) the OS and runtime. The boundary shifts up the stack as you move from IaaS → PaaS → SaaS.

---

**Q17.** Explain the difference between IaaS, PaaS, FaaS, and SaaS.
**A:** IaaS (e.g., EC2) — you manage OS, runtime, app, data; provider manages hardware. PaaS (e.g., Heroku) — you manage app and data; provider manages everything below. FaaS (e.g., Lambda) — you manage only function code; provider manages runtime and scaling. SaaS (e.g., Gmail) — you manage configuration only.

---

**Q18.** What is the difference between vertical and horizontal scaling?
**A:** Vertical scaling increases the size of a single instance (more CPU/RAM). Horizontal scaling adds more instances behind a load balancer. Horizontal is generally preferred for cloud-native apps as it avoids a single point of failure and is cheaper at scale.

---

**Q19.** What are three characteristics of serverless (FaaS) computing?
**A:** (1) Event-driven — functions trigger on HTTP requests, queue messages, file uploads, etc. (2) Stateless — each invocation is independent. (3) Billed per execution — pay for duration and memory used, not idle time.

---

**Q20.** What is a "cold start" in serverless computing?
**A:** When a serverless function hasn't been invoked recently, the cloud platform needs to spin up a new execution environment, introducing latency (milliseconds to seconds). This doesn't happen with pre-warmed containers or VM-based workloads.

---

**Q21.** What is Infrastructure as Code (IaC) and why is it beneficial?
**A:** IaC defines and provisions infrastructure using code (e.g., Terraform HCL, CloudFormation YAML) instead of manual console clicks. Benefits: version-controlled changes, reproducible environments, automated deployments, and auditable history.

---

**Q22.** What is the difference between object storage, block storage, and file storage in the cloud?
**A:** Object storage (e.g., S3) stores flat files/blobs accessed by a key — best for images, backups, and static assets. Block storage (e.g., EBS) provides low-latency disk volumes, best for VM disks and databases. File storage (e.g., EFS) provides a shared NFS/SMB filesystem, best for shared configurations and media.

---

**Q23.** What does the Least Privilege principle mean in IAM?
**A:** Grant only the permissions required for a specific task — no more. Users, roles, and services should not have blanket admin access. This minimizes the blast radius of a compromised account or misconfigured service.

---

**Q24.** What is the difference between AWS SQS (queue) and SNS (pub/sub)?
**A:** SQS is a point-to-point queue — one consumer processes each message. SNS is a publish/subscribe system — one message fans out to multiple subscribers simultaneously. SQS decouples producers from consumers; SNS enables broadcast notifications.

---

**Q25.** What is the purpose of CloudTrail in AWS?
**A:** CloudTrail records an audit log of all API calls made in an AWS account, capturing who did what, when, and from where. It is essential for security monitoring, incident investigation, and compliance.

---

**Q26.** Compare Serverless vs Serverfull (VMs/containers) across three dimensions.
**A:** (1) Management — serverless: none (provider handles all); serverfull: you manage OS, patches, scaling. (2) Cost — serverless: pay per invocation; serverfull: pay per running instance. (3) Cold starts — serverless: yes (latency on first call); serverfull: no (always-running).

---

**Q27.** What is the purpose of a multi-AZ deployment in AWS RDS?
**A:** Multi-AZ creates a synchronous standby replica in a different Availability Zone. If the primary fails, RDS automatically failovers to the standby with minimal downtime, providing high availability without manual intervention.

---

## Part 3 — Software Design

**Q28.** What is the difference between the Layered and Microservices architectural patterns?
**A:** Layered divides the system horizontally (Presentation → Business → Data) within a single deployable unit — good for cohesive apps with clear separation. Microservices decompose by business domain into small, independently deployable services — good for teams needing independent scaling and deployment, at the cost of distributed system complexity.

---

**Q29.** Explain the Circuit Breaker pattern. Why is it important in microservices?
**A:** A Circuit Breaker wraps calls to a remote service. After a threshold of failures, it "opens" and immediately returns an error without calling the downstream service, preventing cascading failures. After a timeout, it enters "half-open" state to probe recovery. It's critical in microservices where one failing service can bring down many others.

---

**Q30.** What is the Saga pattern and when would you use it?
**A:** The Saga pattern manages distributed transactions across multiple microservices, where a single ACID transaction is impossible. It sequences local transactions per service; if one fails, compensating transactions undo previous steps. Used in e-commerce (order, payment, inventory) and any multi-service workflow requiring consistency without 2-phase commit.

---

**Q31.** What are the SOLID principles? Give a one-sentence definition for each.
**A:**
- **S** — Single Responsibility: a class has one reason to change.
- **O** — Open/Closed: open for extension, closed for modification.
- **L** — Liskov Substitution: subtypes must be substitutable for their base type without breaking behaviour.
- **I** — Interface Segregation: prefer many small, specific interfaces over one large general one.
- **D** — Dependency Inversion: depend on abstractions, not concretions.

---

**Q32.** What is Dependency Injection and what problem does it solve?
**A:** DI is the practice of supplying an object's dependencies from outside rather than having it create them internally. It decouples classes from their concrete implementations, making code easier to test (swap real deps with mocks) and easier to change (swap implementations without touching the class).

---

**Q33.** What is the difference between KISS, DRY, and YAGNI?
**A:** KISS — Keep It Simple: avoid unnecessary complexity. DRY — Don't Repeat Yourself: have a single source of truth for every piece of logic. YAGNI — You Aren't Gonna Need It: don't build features for hypothetical future requirements. All three reduce unnecessary code and complexity.

---

**Q34.** What is the Observer design pattern? Give a real-world example.
**A:** Observer defines a one-to-many dependency so that when one object changes state, all its dependents are notified automatically. Example: a UI event system where button clicks (subject) notify multiple event listeners (observers), such as a logging handler, an animation handler, and a state update handler.

---

**Q35.** What is the difference between Functional Programming and Object-Oriented Programming?
**A:** OOP organises code around objects with state and behaviour — good for modelling real-world entities. FP treats computation as pure functions without shared mutable state — good for data transformation, concurrency, and testability. OOP can lead to complex inheritance hierarchies; FP has a steeper learning curve but is more predictable and parallelizable.

---

**Q36.** What is an Architecture Decision Record (ADR)?
**A:** An ADR is a short document recording an architectural decision with its context, the options considered, and the rationale for the choice. It prevents future teams from revisiting the same decision without context (e.g., "Why did we choose Kafka over RabbitMQ?").

---

**Q37.** What is the purpose of the Repository pattern?
**A:** The Repository pattern abstracts the data access layer behind an interface, so application logic doesn't depend directly on database or ORM details. It makes data access swappable (e.g., switch from SQL to NoSQL) and testable (replace with an in-memory repo during tests).

---

**Q38.** What is the difference between structured and unstructured logging? Which is preferred and why?
**A:** Unstructured logging outputs plain text (e.g., `"Error on line 42"`). Structured logging outputs key-value pairs in a machine-readable format like JSON (e.g., `{"level":"ERROR","line":42,"user_id":"u1"}`). Structured logging is preferred because it can be queried, filtered, and aggregated by log management tools (e.g., Elasticsearch, Splunk, CloudWatch Insights).

---

**Q39.** What is the C4 model in software documentation?
**A:** C4 is a hierarchical approach to architecture diagrams with four levels of zoom: Context (system in its environment), Containers (apps and data stores), Components (internal structure of a container), and Code (class-level detail). It helps communicate architecture to different audiences at the right level of detail.

---

## Part 4 — Software Engineering Processes

**Q40.** What are the seven major stages of the SDLC?
**A:** (1) Planning, (2) Requirements, (3) Design, (4) Implementation, (5) Testing, (6) Deployment, (7) Maintenance.

---

**Q41.** What is S-SDLC and why does it exist?
**A:** Secure SDLC embeds security practices into *every phase* of the SDLC rather than bolting security on at the end. It reduces vulnerabilities, minimises breach risk, and ensures compliance (e.g., GDPR, OWASP). Example practices: threat modelling in design, SAST in implementation, pen testing in testing.

---

**Q42.** What is the difference between Scrum and Kanban?
**A:** Scrum uses fixed-length time-boxed sprints (1–4 weeks) with defined roles (PO, SM, Dev Team) and ceremonies (planning, daily standup, review, retro). Kanban is flow-based with no fixed iterations — work is pulled as capacity allows, visualised on a board with WIP limits. Scrum gives predictable cadence; Kanban is more flexible and suits support/maintenance work.

---

**Q43.** What are the four Scrum ceremonies?
**A:** Sprint Planning, Daily Standup, Sprint Review, and Sprint Retrospective.

---

**Q44.** What is the difference between a functional requirement and a non-functional requirement (NFR)? Give an example of each.
**A:** A functional requirement defines *what* the system does (e.g., "A user can reset their password via email"). An NFR defines *how well* the system performs — a quality attribute (e.g., "The login page loads in under 2 seconds for 95% of requests"). NFRs cover Performance, Security, Scalability, Reliability, Usability, and Maintainability.

---

**Q45.** What are Acceptance Criteria and what format are they commonly written in?
**A:** Acceptance Criteria define the conditions a feature must meet to be accepted by the stakeholder. They are commonly written in Given/When/Then (Gherkin) format: "Given valid credentials → user is redirected to dashboard."

---

**Q46.** What is the difference between Definition of Ready (DoR) and Definition of Done (DoD)?
**A:** DoR ensures a task is ready to be worked on (story written, AC defined, designs available). DoD ensures the task is truly complete (code reviewed, tests passing, deployed to staging, docs updated). Both prevent ambiguity and maintain team quality standards.

---

**Q47.** What is Planning Poker and what problem does it solve?
**A:** Planning Poker is an estimation technique where each team member simultaneously reveals a card with their effort estimate (Fibonacci sequence). Outliers are discussed, revealing differing assumptions. It solves anchoring bias (where the first number heard influences all others) and builds shared understanding of scope.

---

**Q48.** What is velocity in Agile and how is it used?
**A:** Velocity is the average number of story points a team completes per sprint. It is used for forecasting — if velocity is 40 points/sprint and the backlog has 200 points, the team can forecast ~5 sprints to completion. Stable velocity enables reliable sprint planning.

---

**Q49.** What is the difference between cycle time and lead time?
**A:** Cycle time is the time from when work *starts* on a task to when it is complete. Lead time is the time from when a task is *requested* (entered the backlog) to when it is delivered. Lead time includes waiting time; cycle time only measures active work.

---

**Q50.** What is STRIDE in the context of threat modelling?
**A:** STRIDE is a framework for categorising security threats: **S**poofing (impersonating users/systems), **T**ampering (modifying data), **R**epudiation (denying actions), **I**nformation Disclosure (data leaks), **D**enial of Service (making service unavailable), **E**levation of Privilege (gaining unauthorised access). Used during design to systematically identify risks.

---

**Q51.** How would you approach estimating a task with significant unknowns?
**A:** (1) Time-box a spike (research task) to reduce uncertainty. (2) Use range estimates ("3–7 days"). (3) Apply a risk buffer (20–30%). (4) Break it into smaller pieces and estimate each. (5) Reassess after the first iteration with real data.

---

## Part 5 — Engineering Knowledge & Experience

**Q52.** What is the time complexity of binary search and what precondition does the data need?
**A:** O(log n). The data must be sorted. Binary search halves the search space on each step by comparing the target with the midpoint element.

---

**Q53.** What is the difference between a clustered and a non-clustered index?
**A:** A clustered index defines the physical row order on disk — there can only be one per table (usually the PK). A non-clustered index is a separate data structure that stores pointers to rows — a table can have many. Non-clustered indexes speed up specific queries but add write overhead.

---

**Q54.** What is symmetric vs asymmetric encryption? Give an example of each.
**A:** Symmetric — same key encrypts and decrypts (e.g., AES-256). Fast; used for bulk data encryption. Asymmetric — a public key encrypts, a private key decrypts (or vice versa for signatures) (e.g., RSA, ECDSA). Slower; used for key exchange and digital signatures (e.g., TLS handshake).

---

**Q55.** What are the ACID properties of a database transaction?
**A:** **Atomicity** — all operations in the transaction succeed or all are rolled back. **Consistency** — the DB moves from one valid state to another, respecting all constraints. **Isolation** — concurrent transactions don't see each other's intermediate states. **Durability** — once committed, data survives crashes (persisted to disk/WAL).

---

**Q56.** What is the difference between a VM and a container?
**A:** A VM includes a full OS kernel and runs on a hypervisor — heavyweight, GBs, minutes to start, strong isolation. A container shares the host OS kernel using namespaces and cgroups — lightweight, MBs, seconds to start. VMs offer stronger isolation; containers are more efficient and faster to deploy.

---

**Q57.** Describe the Blue/Green deployment strategy and its main trade-off.
**A:** Maintain two identical environments (Blue = current production, Green = new version). Deploy to Green, run tests, then switch all traffic from Blue to Green. Instant rollback by switching back to Blue. Trade-off: requires double the infrastructure during the transition.

---

**Q58.** What is a Canary deployment?
**A:** Route a small percentage of traffic (e.g., 5%) to the new version while the rest goes to the old. Monitor for errors; gradually increase traffic to the new version. Low risk — only a small subset of users is affected if something goes wrong. Requires more complex traffic routing.

---

**Q59.** What causes a deadlock in multithreading and how can it be prevented?
**A:** A deadlock occurs when two or more threads each hold a lock the other needs, causing all to wait indefinitely. Prevention: always acquire multiple locks in the same consistent order across all threads, use try-lock with timeout, or redesign to avoid needing multiple locks simultaneously.

---

**Q60.** What is the difference between Type 1 and Type 2 hypervisors?
**A:** Type 1 (bare-metal) runs directly on physical hardware (e.g., VMware ESXi, KVM) — higher performance, used in production. Type 2 (hosted) runs on top of a host OS (e.g., VirtualBox, VMware Workstation) — easier setup, used for development and testing.

---

**Q61.** What software license type requires derivative works to also be open-sourced?
**A:** Copyleft licenses, specifically GPL (v2/v3). Any software that uses or derives from GPL-licensed code must also be released under the GPL. LGPL allows dynamic linking in proprietary code without triggering this requirement.

---

**Q62.** What are the OSI model's application, transport, and network layers, with one protocol example each?
**A:** Application layer — HTTP, DNS. Transport layer — TCP, UDP. Network layer — IP, ICMP.

---

## Part 6 — Python OOP

**Q63.** What are the four pillars of OOP?
**A:** Encapsulation (bundling data and methods, hiding internal state), Abstraction (exposing only necessary interfaces), Inheritance (child class derives from parent), Polymorphism (different classes respond to the same interface differently).

---

**Q64.** What is the difference between a class attribute and an instance attribute in Python?
**A:** A class attribute is defined on the class itself and shared across all instances. An instance attribute is defined via `self` in `__init__` and is unique per object. If you assign to an instance attribute with the same name as a class attribute, it creates a new instance attribute that *shadows* the class attribute for that instance only.

---

**Q65.** What is the difference between `__repr__` and `__str__`?
**A:** `__repr__` is the unambiguous developer representation, used by `repr()`, the debugger, and the REPL. `__str__` is the human-readable representation, used by `print()` and `str()`. If `__str__` is absent, Python falls back to `__repr__`. Rule of thumb: always implement `__repr__`.

---

**Q66.** What is the `@property` decorator used for?
**A:** It allows a method to be accessed as if it were an attribute (no parentheses). Combined with `@property.setter`, it provides controlled getters and setters — enabling validation, computed values, and encapsulation without changing the calling interface.

---

**Q67.** What is `__slots__` and when should you use it?
**A:** `__slots__` restricts instance attributes to a declared set, preventing creation of a per-instance `__dict__`. This reduces memory usage and speeds up attribute access. Use it when creating many instances (e.g., game entities, data records) where memory matters.

---

**Q68.** What is the MRO (Method Resolution Order) and what algorithm does Python use?
**A:** MRO defines the order Python searches for methods in an inheritance hierarchy. Python uses **C3 Linearization** — a deterministic left-to-right algorithm that ensures each class appears only once and respects the order of parent classes. Inspect with `ClassName.__mro__`.

---

**Q69.** What is the diamond problem in OOP and how does Python resolve it?
**A:** The diamond problem arises when class D inherits from B and C, both of which inherit from A. Without a resolution rule, calling `D.method()` is ambiguous. Python resolves it with C3 Linearization, producing a consistent MRO (D → B → C → A → object) that follows the order classes are declared in `D(B, C)`.

---

**Q70.** What is a Mixin in Python?
**A:** A Mixin is a small, focused class that adds a specific capability (e.g., JSON serialisation, logging) and is not meant to be instantiated alone. Combined via multiple inheritance, they let you compose behaviour without deep inheritance hierarchies (e.g., `class User(JsonMixin, LogMixin)`).

---

**Q71.** What is the difference between composition and inheritance? When should you prefer composition?
**A:** Inheritance models "is-a" relationships — tight coupling, reuse via subclassing. Composition models "has-a" relationships — loose coupling, reuse via delegation. Prefer composition when the relationship is "has-a", when behaviour needs to change at runtime, or when deep inheritance hierarchies are becoming hard to manage.

---

**Q72.** What is the difference between an Association, Aggregation, and Composition relationship?
**A:** Association — objects use each other but exist independently (e.g., Driver ↔ Car). Aggregation — parent owns child but child can exist without parent (e.g., Team has Players). Composition — parent owns child and child cannot exist without parent (e.g., House has Rooms — rooms are created and destroyed with the house).

---

**Q73.** What does `@classmethod` do and how does it differ from `@staticmethod`?
**A:** `@classmethod` receives the class (`cls`) as its first argument. Used for alternative constructors or class-level operations. `@staticmethod` receives no implicit first argument — it's a utility function grouped in the class but needing neither `self` nor `cls`.

---

**Q74.** How does Python implement "private" attributes, and are they truly private?
**A:** Single underscore `_name` is a convention meaning "internal use" — not enforced. Double underscore `__name` triggers name-mangling, storing it as `_ClassName__name`. It's not truly private — you can still access it via the mangled name — but it prevents accidental overriding in subclasses.

---

**Q75.** What is an Abstract Base Class (ABC) in Python and how do you create one?
**A:** An ABC enforces that subclasses implement specific methods. Use `from abc import ABC, abstractmethod`, inherit from `ABC`, and decorate unimplemented methods with `@abstractmethod`. Attempting to instantiate a class with unimplemented abstract methods raises `TypeError`.

---

**Q76.** What is `__hash__` and why must it be consistent with `__eq__`?
**A:** `__hash__` returns an integer used to place an object in hash-based collections (sets, dict keys). If two objects are equal (`__eq__` returns True), they *must* have the same hash. Python sets `__hash__ = None` if you define `__eq__` without `__hash__`, making the object unhashable.

---

## Part 7 — Python Functions

**Q77.** What is the LEGB rule in Python scoping?
**A:** Python resolves names in this order: **L**ocal (current function), **E**nclosing (outer functions), **G**lobal (module level), **B**uilt-in (Python built-ins like `len`, `print`). The first match is used.

---

**Q78.** What is a closure in Python? Give an example.
**A:** A closure is an inner function that *captures* variables from its enclosing scope, even after the outer function has returned. Example:
```python
def multiplier(factor):
    def multiply(x):
        return x * factor  # 'factor' is captured
    return multiply
double = multiplier(2)
double(5)  # 10
```

---

**Q79.** What is the difference between `global` and `nonlocal` keywords?
**A:** `global` declares that a variable refers to the module-level (global) name, allowing you to modify it inside a function. `nonlocal` declares that a variable refers to the nearest enclosing (non-global) scope, allowing you to modify it from an inner function.

---

**Q80.** What happens to an exception variable after a `try/except` block in Python 3?
**A:** The exception variable is deleted from scope after the `except` block completes — even if it was defined before. This is intentional to avoid reference cycles. If you need the exception after the block, save it to a separate variable inside the `except` block.

---

**Q81.** What does `functools.reduce` do? Give an example.
**A:** `reduce(func, iterable)` accumulates a sequence down to a single value by repeatedly applying a binary function. Example: `reduce(lambda a, b: a + b, [1, 2, 3, 4])` returns `10` (((1+2)+3)+4).

---

**Q82.** What is a higher-order function? Name two examples from Python's standard library.
**A:** A higher-order function either takes a function as an argument or returns one. Examples: `map()` (takes a function and applies it to each element), `filter()` (takes a predicate and keeps matching elements), `sorted(key=...)` (takes a key function), `functools.partial`.

---

**Q83.** Can `globals()` be mutated to create new global variables? Can `locals()` be mutated?
**A:** Yes — `globals()` returns the actual global namespace dict; mutations affect real global variables. No — `locals()` returns a *snapshot copy* inside functions; modifying it does not change local variables.

---

**Q84.** What are the two requirements every recursive function must have?
**A:** (1) A **base case** (termination condition) that stops recursion. (2) A **recursive step** that reduces the problem toward the base case. Without a base case, recursion continues until a `RecursionError` (stack overflow).

---

**Q85.** What is the default Python recursion limit and how would you change it?
**A:** The default is approximately 1000 calls. You can change it with `sys.setrecursionlimit(n)`. However, deep recursion is generally avoided in Python because there is no tail-call optimization (TCO) — iterative solutions are preferred for performance.

---

## Part 8 — Python Exceptions

**Q86.** What is the difference between `Exception` and `BaseException`?
**A:** `BaseException` is the root of all exceptions and includes `SystemExit`, `KeyboardInterrupt`, and `GeneratorExit`. `Exception` is a subclass of `BaseException` and covers "normal" application exceptions. You should always inherit custom exceptions from `Exception`, not `BaseException`, to avoid accidentally suppressing program exit or Ctrl+C.

---

**Q87.** What is the purpose of the `else` clause in a `try/except` block?
**A:** The `else` clause runs only if *no exception was raised* in the `try` block. It's useful for code that should only execute on success, keeping it separate from the `try` block to avoid accidentally catching exceptions from the success path.

---

**Q88.** What is the purpose of `finally` in a `try` block?
**A:** `finally` always executes, regardless of whether an exception was raised, caught, or re-raised — and even if a `return` is encountered. Used for cleanup operations like closing files, releasing locks, or closing database connections.

---

**Q89.** What is exception chaining? What does `raise X from Y` do?
**A:** Exception chaining preserves the original exception context when raising a new one. `raise RuntimeError("msg") from original_error` explicitly chains them, so the traceback shows both. `raise X from None` suppresses the original context — useful when you want to show only the high-level error.

---

**Q90.** Why should you *not* use `assert` for data validation in production code?
**A:** `assert` statements are disabled when Python is run with the `-O` (optimise) flag (`python -O script.py`), meaning they are silently skipped. Use explicit `if`/`raise` instead for production validation. `assert` is a debugging tool only.

---

**Q91.** How would you create a custom exception that carries additional data?
**A:** Subclass `Exception` and override `__init__` to store the extra data as instance attributes, then call `super().__init__(message)`. Example:
```python
class InsufficientFundsError(Exception):
    def __init__(self, amount, balance):
        self.amount = amount
        self.balance = balance
        super().__init__(f"Cannot withdraw {amount}. Balance: {balance}.")
```

## Part 9 — Python Testing

**Q92.** What is the difference between a Mock, a Stub, and a Patch in Python testing?
**A:** A **Mock** replaces an object and records calls — used to verify interactions (was a method called with correct args?). A **Stub** returns predefined data — used to control indirect inputs to the unit under test. A **Patch** temporarily replaces something in scope (e.g., `requests.get`) to avoid side effects during the test.

---

**Q93.** What is TDD (Test-Driven Development) and what is its core cycle?
**A:** TDD is the practice of writing a failing test *before* writing the implementation. The cycle is Red → Green → Refactor: (1) Write a failing test, (2) Write the minimum code to pass it, (3) Refactor while keeping tests green.

---

**Q94.** What causes test flakiness and how do you address it?
**A:** Causes: time dependencies, shared/global state, race conditions, real network calls. Solutions: freeze time with `freezegun`, isolate test state (fresh DB per test), replace network calls with mocks or the `responses` library, use deterministic data.

---

**Q95.** What is the key difference between `pytest` and `unittest`?
**A:** `pytest` uses simple function-based tests with plain `assert` statements — minimal boilerplate, powerful fixtures, rich plugin ecosystem. `unittest` is class-based, built-in, uses `assertEqual`/`assertRaises` style assertions. pytest is generally preferred for new projects; unittest is used with legacy code or frameworks that require it.

---

**Q96.** What is the difference between line coverage and branch coverage?
**A:** Line coverage measures which lines were executed during tests. Branch coverage measures whether all `if/else` paths (branches) were taken. Branch coverage is stricter — you can have 100% line coverage while missing half the conditional branches.

---

**Q97.** What is the N+1 query problem and how is it fixed?
**A:** When you load a list of objects and then access a related object on each in a loop, it fires one additional query per row (N+1 total instead of 1). Fixed in Django by using `select_related()` (JOIN for ForeignKey) or `prefetch_related()` (separate prefetch for ManyToMany).

---

**Q98.** Why should you not mock what you own (the "mock at the boundary" rule)?
**A:** Mocking internal/owned code tightly couples tests to implementation details. If you refactor the internal implementation, tests break even though behaviour hasn't changed. You should only mock at system boundaries — external HTTP calls, databases, file I/O, third-party services — where you have no control and want to avoid real side effects.

---

**Q99.** What is `@pytest.mark.parametrize` used for?
**A:** It allows you to run the same test function with multiple sets of inputs, eliminating duplicated test code. Example:
```python
@pytest.mark.parametrize("a,b,expected", [(1,2,3),(0,0,0),(-1,1,0)])
def test_add(a, b, expected):
    assert add(a, b) == expected
```

## Part 10 — Python Async & Parallel

**Q100.** What is the difference between concurrency and parallelism?
**A:** Concurrency means *dealing with* multiple tasks at once — tasks may interleave but don't necessarily run at the same physical instant (single core switching). Parallelism means *executing* multiple tasks at the exact same instant on multiple CPU cores. In Python: `asyncio`/`threading` = concurrency; `multiprocessing` = parallelism.

---

**Q101.** What is the GIL and what impact does it have on Python threading?
**A:** The Global Interpreter Lock is a mutex in CPython that allows only one thread to execute Python bytecode at a time. It exists to protect reference counting (memory management). For I/O-bound tasks, threads release the GIL while waiting, so threading works well. For CPU-bound tasks, the GIL prevents true parallelism — use `multiprocessing` instead.

---

**Q102.** When should you use `asyncio` vs `threading` vs `multiprocessing`?
**A:** `asyncio` — I/O-bound tasks with async-compatible libraries (network, APIs). `ThreadPoolExecutor` / `threading` — I/O-bound tasks with blocking libraries that can't use async. `ProcessPoolExecutor` / `multiprocessing` — CPU-bound tasks that need true parallelism (each process has its own GIL).

---

**Q103.** What is the event loop in asyncio and how does it work?
**A:** The event loop is the core of asyncio. It maintains a queue of coroutines and callbacks. It runs a coroutine until it hits an `await`, then picks up another ready task. When the awaited result is ready, it resumes the original coroutine. This cooperative multitasking allows thousands of concurrent tasks on a single thread.

---

**Q104.** What is the difference between a coroutine, a Task, and a Future in asyncio?
**A:** A **coroutine** is defined with `async def` — it does nothing until awaited or scheduled. A **Task** wraps a coroutine and schedules it to run concurrently on the event loop (`asyncio.create_task()`). A **Future** is a low-level placeholder for a result that hasn't been computed yet; Tasks are a subclass of Future.

---

**Q105.** What does `asyncio.gather()` do?
**A:** It runs multiple coroutines concurrently and waits for all to complete, returning results in the same order as the inputs. For example, three tasks that each take 1 second complete in ~1 second total (not 3 seconds) because they run concurrently.

---

**Q106.** What is a race condition? Provide a Python threading example.
**A:** A race condition occurs when the result depends on the unpredictable interleaving of threads accessing shared data. Example: two threads both reading a counter, incrementing it, and writing back — the increment is not atomic, so increments are lost. Fix: use a `threading.Lock` around the critical section.

---

**Q107.** What is the difference between a `threading.Lock` and a `threading.Semaphore`?
**A:** A Lock allows only *one* thread into a critical section at a time (mutual exclusion). A Semaphore allows up to *N* threads simultaneously — useful for limiting concurrent access to a pool of resources (e.g., max 3 concurrent DB connections).

---

**Q108.** What is `uvloop` and when should you use it?
**A:** `uvloop` is a drop-in replacement for the default asyncio event loop, built on libuv (the Node.js engine). It can be 2–4x faster than the default loop. Use it for high-throughput servers (REST APIs, WebSockets) on Linux/macOS. It is not available on Windows.

---

**Q109.** What is `contextvars.ContextVar` and why is it preferred over `threading.local()` in async code?
**A:** `ContextVar` provides per-task isolated state in asyncio. `threading.local()` gives per-thread storage, but asyncio runs many tasks on a single thread, so all tasks would share the same thread-local value. `ContextVar` correctly scopes values to each individual asyncio Task, making it the right tool for request-scoped data (e.g., user ID, trace ID) in async web frameworks.

---

**Q110.** What is Cooperative vs Preemptive multitasking? Which does asyncio use?
**A:** Preemptive — the OS can interrupt a running thread at any point and switch to another. Cooperative — a task runs until it voluntarily yields control (at `await` or `yield`). asyncio uses cooperative multitasking. The implication: one blocking call (e.g., `time.sleep()` instead of `asyncio.sleep()`) stalls the entire event loop, starving all other tasks.

## Part 11 — Python ORM & Databases

**Q111.** What is the difference between the Active Record and Data Mapper patterns?
**A:** Active Record — the model class IS the table; instances know how to save/update/delete themselves (e.g., Django ORM's `user.save()`). Data Mapper — the model is a plain object; a separate mapper (session) handles all DB operations (e.g., SQLAlchemy's `session.add(user); session.commit()`). Active Record is simpler; Data Mapper is more flexible and decoupled.

---

**Q112.** What is an ORM migration and why is it necessary?
**A:** A migration is a version-controlled script that evolves the database schema (adding tables, renaming columns, creating indexes) to match updated models. Migrations are necessary to apply schema changes consistently across all environments (dev, staging, production) without losing existing data, and to enable rollback.

---

**Q113.** What is the difference between Django's `select_related` and `prefetch_related`?
**A:** `select_related` performs a SQL JOIN for ForeignKey/OneToOne relationships — fetches everything in one query. `prefetch_related` performs separate queries for ManyToMany or reverse ForeignKey relationships and does the joining in Python. Use `select_related` for "to-one" relations and `prefetch_related` for "to-many".

---

**Q114.** What are Django `F` objects used for?
**A:** `F` objects reference a field's value in the database without pulling it into Python. This enables atomic DB-level operations. Example: `Product.objects.update(price=F("price") + 10)` updates all prices in a single SQL `UPDATE` without a round-trip to Python, avoiding race conditions.

---

**Q115.** What is a database view and what is the difference between a view and a materialized view?
**A:** A view is a saved query stored as a virtual table — it re-runs the query on every access and stores no data. A materialized view physically caches the query result and must be refreshed periodically. Materialized views are faster for read-heavy reporting queries but show potentially stale data.

---

**Q116.** What is the difference between `UNION` and `UNION ALL`?
**A:** `UNION` combines result sets and removes duplicate rows (requires additional sorting/deduplication). `UNION ALL` keeps all rows including duplicates and is faster. Use `UNION ALL` when duplicates are either impossible or acceptable.

---

**Q117.** What are window functions in SQL? Give an example.
**A:** Window functions perform calculations across a set of related rows *without collapsing them* (unlike GROUP BY). They use `OVER(PARTITION BY ... ORDER BY ...)`. Example: `RANK() OVER (PARTITION BY dept_id ORDER BY salary DESC)` ranks employees by salary within each department without losing individual rows.

---

**Q118.** What is database normalization? Describe 1NF, 2NF, and 3NF.
**A:** Normalization organises tables to reduce redundancy. **1NF** — all values are atomic (no arrays), each row is unique. **2NF** — 1NF + no partial dependency on a composite PK (every non-key column depends on the whole PK). **3NF** — 2NF + no transitive dependency (non-key column depending on another non-key column).

---

**Q119.** What is the CAP theorem?
**A:** In a distributed database, you can only guarantee two of three properties: **Consistency** (every read returns the most recent write), **Availability** (every request gets a response), **Partition Tolerance** (system works despite network failures). Since partitions are unavoidable, the real trade-off is CP (consistent, may reject requests during partition) vs AP (always responds, may return stale data).

---

**Q120.** What are transaction isolation levels and what anomalies does each prevent?
**A:** Four standard levels: **READ UNCOMMITTED** — prevents nothing (dirty reads possible). **READ COMMITTED** — prevents dirty reads. **REPEATABLE READ** — prevents dirty reads and non-repeatable reads. **SERIALIZABLE** — prevents all anomalies including phantom reads. Higher isolation means fewer anomalies but more locking and reduced concurrency.
