## Section 1: Time Complexity & Big O Notation

**Q1.** What is the time complexity of the following code and why?

```python
n = 100
x = n * 2
```

<details>
<summary>Answer</summary>

**O(1) — Constant time.** The operation executes once regardless of the value of `n`. Assignment and arithmetic are single operations.

</details>

---

**Q2.** What is the time complexity of this loop?

```python
for i in range(n):
    print(i)
```

<details>
<summary>Answer</summary>

**O(n) — Linear time.** The `print` statement executes exactly `n` times, so the runtime grows linearly with `n`.

</details>

---

**Q3.** What is the time complexity of the following loop, and does changing the step size affect the Big O?

```python
for i in range(0, n, 5):
    print(i)
```

<details>
<summary>Answer</summary>

**O(n) — Still linear.** The loop executes `n/5` times, but the constant `1/5` is dropped in Big O notation. Changing the step by a **constant** does not change the complexity class.

</details>

---

**Q4.** What is the time complexity of this `while` loop and why is it different from a linear loop?

```python
n = 100
while n > 0:
    n = n // 2
```

<details>
<summary>Answer</summary>

**O(log n) — Logarithmic time.** The value of `n` is halved on each iteration. The number of iterations is approximately log₂(n), not n. When a loop counter is multiplied or divided by a constant factor, the complexity is logarithmic.

</details>

---

**Q5.** What is the time complexity of this nested loop?

```python
for i in range(n):
    for j in range(n):
        print(i, j)
```

<details>
<summary>Answer</summary>

**O(n²) — Quadratic time.** The inner loop runs `n` times for each iteration of the outer loop, giving `n × n = n²` total operations.

</details>

---

**Q6.** Rank these complexities from fastest to slowest: O(n²), O(1), O(n log n), O(log n), O(n).

<details>
<summary>Answer</summary>

**O(1) < O(log n) < O(n) < O(n log n) < O(n²)**

- O(1): Constant
- O(log n): Logarithmic
- O(n): Linear
- O(n log n): Linearithmic
- O(n²): Quadratic

</details>

## Section 2: Linear Search

**Q7.** What does the following function return if the target is not found in `data`?

```python
def linear_search(data, target):
    for idx, val in enumerate(data):
        if val == target:
            return idx
    return -1
```

<details>
<summary>Answer</summary>

**-1.** The function returns `-1` as a sentinel value indicating the target was not found in the list.

</details>

## Section 3: Selection Sort

**Q10.** Trace through the selection sort algorithm with `xs = [3, 2, 1]`. What are the values after each pass of the outer loop?

```python
def selection_sort(xs):
    for i in range(len(xs)-1):
        min_idx = i
        for j in range(i+1, len(xs)):
            if xs[j] < xs[min_idx]:
                min_idx = j
        xs[i], xs[min_idx] = xs[min_idx], xs[i]
    return xs
```

<details>
<summary>Answer</summary>

- Initial: `[3, 2, 1]`
- After pass 1 (i=0): min is `1` at index 2 → swap → `[1, 2, 3]`
- After pass 2 (i=1): min is `2` at index 1 → no swap → `[1, 2, 3]`
- Result: `[1, 2, 3]`

</details>

---

**Q11.** What is the time complexity of selection sort and why?

<details>
<summary>Answer</summary>

**O(n²).** It uses two nested loops: the outer loop runs `n-1` times and the inner loop runs `n-i-1` times for each `i`. Total comparisons: (n-1) + (n-2) + ... + 1 = n(n-1)/2, which is O(n²).

</details>

---

**Q12.** How does selection sort find the minimum element to place in position `i`?

<details>
<summary>Answer</summary>

It sets `min_idx = i` (assuming the current position holds the minimum), then scans the **unsorted portion** of the list (`i+1` to end). If a smaller element is found, `min_idx` is updated. After the inner loop, it **swaps** `xs[i]` with `xs[min_idx]`.

</details>

## Section 4: Greedy Algorithm — Make Change

**Q13.** Given the `make_change` function below, what is the output for `make_change(24)`?

```python
def make_change(target_amount):
    denominations = [200, 100, 50, 20, 10, 5, 2, 1]
    coin_count = 0
    values = []
    for coin in denominations:
        while target_amount >= coin:
            target_amount -= coin
            values.append(coin)
            coin_count += 1
    return coin_count, values
```

<details>
<summary>Answer</summary>

**(3, [20, 2, 2])**

- 24 ≥ 20 → use 20, remainder = 4
- 4 ≥ 2 → use 2, remainder = 2
- 2 ≥ 2 → use 2, remainder = 0
- Total: 3 coins

</details>

---

**Q14.** What strategy does `make_change` use and what are its limitations?

<details>
<summary>Answer</summary>

It uses a **greedy algorithm**: always pick the largest denomination that fits. This is optimal for standard currency systems (like the one with [200, 100, 50, 20, 10, 5, 2, 1]), but **fails** for arbitrary coin sets. For example, with coins `[1, 3, 4]` and target `6`, greedy picks `4+1+1` (3 coins), but the optimal is `3+3` (2 coins).

</details>

## Section 6: The 100 Doors Problem

**Q17.** In the 100 doors problem, 100 doors start closed. On pass `k`, you toggle every `k`th door (starting at door `k`). After 100 passes, which doors are open?

<details>
<summary>Answer</summary>

Doors **1, 4, 9, 16, 25, 36, 49, 64, 81, 100** — the **perfect squares** up to 100.

A door ends open if it is toggled an **odd** number of times. Door `d` is toggled once for each of its divisors. Most numbers have divisors in pairs (e.g., 12: 1×12, 2×6, 3×4), giving an even count. Perfect squares have one **unpaired** divisor (the square root), giving an odd count — so they remain open.

</details>

## Section 7: Palindrome Index

**Q19.** What does the `palindromeIndex` function return for input `"aaab"`?

```python
def palindromeIndex(s: str):
    if s != s[::-1]:
        size = len(s)
        for i in range(0, size):
            straight = s[:i] + s[i+1:size]
            if straight == straight[::-1]:
                return i
    return -1
```

<details>
<summary>Answer</summary>

**3.** The string `"aaab"` is not a palindrome. Removing index 3 (`'b'`) gives `"aaa"`, which is a palindrome. So it returns `3`.

</details>

---

**Q20.** What does `s[::-1]` do in Python?

<details>
<summary>Answer</summary>

It **reverses** the string using slice notation. The syntax `[start:stop:step]` with `step=-1` traverses the string from end to start. `s[::-1]` returns a new string that is the reverse of `s`.

</details>

<details>
<summary>Answer</summary>

**-1.** The first condition `if s != s[::-1]` is False for a palindrome, so the function skips the loop and returns `-1`.

</details>

## Section 9: Python Concepts

**Q24.** What does `itertools.combinations(iterable, r)` return?

```python
from itertools import combinations
perm = combinations([58, 60, 20, 30, 10], 2)
```

<details>
<summary>Answer</summary>

It returns all **unique combinations** of length `r` from the iterable, **without repetition** and **without regard to order**. For 5 elements taken 2 at a time: C(5,2) = 10 combinations. Example: `(58, 60)`, `(58, 20)`, etc.

</details>

---

**Q25.** What is the difference between `combinations` and `permutations` in Python's `itertools`?

<details>
<summary>Answer</summary>

- **`combinations`**: order does NOT matter. `(1, 2)` and `(2, 1)` are the same.
- **`permutations`**: order MATTERS. `(1, 2)` and `(2, 1)` are different.

For `n` items taken `r` at a time:
- Combinations: C(n,r) = n! / (r!(n-r)!)
- Permutations: P(n,r) = n! / (n-r)!

</details>
