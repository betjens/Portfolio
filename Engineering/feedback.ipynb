### SQL vs. NoSQL

**SQL (Relacional)**
- **Características**:
  - Datos estructurados con esquemas fijos.
  - Usan SQL para consultas complejas.
- **Ventajas**:
  - Integridad y consistencia de datos.
  - Ideal para aplicaciones con relaciones complejas.
- **Casos de Uso**:
  - Sistemas de gestión empresarial, finanzas.

**NoSQL (No Relacional)**
- **Características**:
  - Flexibilidad en la estructura de datos, a menudo no requieren esquemas.
  - Escalabilidad horizontal.
- **Ventajas**:
  - Mejor rendimiento con grandes volúmenes de datos no estructurados.
  - Adecuado para aplicaciones en tiempo real.
- **Casos de Uso**:
  - Redes sociales, aplicaciones de IoT, análisis de Big Data.

**Factores de Decisión**
1. **Estructura de Datos**: SQL para datos estructurados, NoSQL para datos no estructurados.
2. **Escalabilidad**: NoSQL ofrece mejor escalabilidad horizontal.
3. **Consistencia**: SQL garantiza ACID; NoSQL puede priorizar la disponibilidad.

### Diferencias entre DROP, DELETE y TRUNCATE en SQL

1. **DROP**:
   - Elimina una tabla o base de datos completamente, incluyendo su estructura y datos.
   - No se puede recuperar la tabla eliminada.

2. **DELETE**:
   - Elimina filas específicas de una tabla, basándose en una condición.
   - Los datos se pueden recuperar si hay una transacción activa.

3. **TRUNCATE**:
   - Elimina todos los registros de una tabla de manera rápida, pero conserva la estructura.
   - No se puede usar con una cláusula WHERE y no registra cada fila eliminada. 

### Joins

**1. Inner Join**: Devuelve los registros con valores coincidentes en ambas tablas. Excluye las filas que no coinciden.

**2. Left Join (Left Outer Join)**: Devuelve todos los registros de la tabla de la izquierda y los registros coincidentes de la tabla de la derecha. Si no hay coincidencia, se devuelven valores NULL para la tabla de la derecha.

**3. Right Join (Right Outer Join)**: Devuelve todos los registros de la tabla de la derecha y los registros coincidentes de la tabla de la izquierda. Se devuelven NULLs para los registros no coincidentes en la tabla de la izquierda.

**4. Full Join (Full Outer Join)**: Combina los resultados de los joins izquierdo y derecho. Devuelve todos los registros cuando hay coincidencia en cualquiera de las tablas.

**5. Cross Join**: Devuelve el producto cartesiano de ambas tablas, combinando cada fila de la primera tabla con cada fila de la segunda.

### Soluciones de bases de datos NoSQL:

### Gestión de memoria

La administración de memoria (memory management) es fundamental para optimizar el rendimiento y evitar problemas en aplicaciones que usan muchos recursos o que se ejecutan por largo tiempo. En lenguajes como Python, esta tarea se gestiona principalmente de forma automática gracias al **recolector de basura (garbage collector)**, pero entender cómo funciona y cómo manejarla eficientemente ayuda a los desarrolladores a escribir código más eficiente y seguro.

Aquí tienes algunos conceptos clave sobre la administración de memoria, particularmente en Python:

#### 1. **Asignación de Memoria y Estructuras de Datos**
   - Python utiliza una capa de abstracción para manejar la memoria, lo que hace que sea más fácil para los desarrolladores trabajar sin preocuparse demasiado por detalles específicos de hardware.
   - Cada objeto creado en Python ocupa memoria, y el tamaño depende del tipo de datos. Las estructuras de datos como listas, diccionarios y conjuntos pueden ocupar más espacio, especialmente si contienen otros objetos complejos.
   - Python administra internamente su memoria en bloques, manejando pequeños y grandes bloques de manera diferente para optimizar la velocidad y reducir el consumo.

#### 2. **Contador de Referencias**
   - Python utiliza un contador de referencias para rastrear cuántas veces se usa un objeto. Cuando este contador llega a cero (es decir, el objeto ya no está referenciado), el recolector de basura lo elimina automáticamente, liberando memoria.
   - Un inconveniente del contador de referencias es que no puede manejar referencias cíclicas (por ejemplo, si el objeto A referencia a B y B a A), lo que puede llevar a fugas de memoria si no se manejan adecuadamente.

#### 3. **Recolector de Basura (Garbage Collector)**
   - El recolector de basura de Python complementa el sistema de referencia para gestionar las referencias cíclicas. Utiliza una técnica llamada **recolección generacional**, que organiza los objetos en generaciones según su tiempo de vida y frecuencia de uso.
   - El recolector escanea y elimina los objetos no referenciados en ciclos periódicos, siendo más agresivo con los objetos de vida corta (objetos jóvenes), lo que optimiza el rendimiento en general.
   
#### 4. **Memoria Dinámica y Fragmentación**
   - Python maneja la memoria dinámica mediante la biblioteca de C **malloc** y **free**, pero también tiene su propio sistema para manejar estructuras de datos pequeñas y de uso frecuente.
   - La fragmentación puede ser un problema, especialmente en aplicaciones que crean y destruyen grandes cantidades de objetos en memoria, lo que deja "huecos" en la memoria y puede disminuir la eficiencia.
   - Para mitigar este problema, el recolector de basura reorganiza internamente los bloques de memoria, lo que ayuda a reducir la fragmentación.

#### 5. **Optimización y Herramientas**
   - Los desarrolladores pueden monitorear y optimizar el uso de memoria utilizando herramientas como **memory_profiler**, **objgraph**, y **tracemalloc**, que ofrecen estadísticas sobre el uso de memoria y ayudan a identificar posibles problemas.
   - Evitar referencias cíclicas explícitas, reutilizar objetos y emplear estructuras de datos de tamaño adecuado también son buenas prácticas que optimizan el uso de la memoria.

### [multithreading vs multiprocessing](https://realpython.com/python-concurrency/#what-is-concurrency)

#### ¿Qué es la Concurrencia?

La concurrencia se refiere a la capacidad de un sistema para gestionar múltiples tareas al mismo tiempo. No significa necesariamente que las tareas se ejecuten simultáneamente, sino que el sistema puede gestionar y alternar entre varias tareas para hacerlas avanzar de forma aparentemente simultánea.

En Python, la concurrencia es especialmente útil cuando se trabaja con tareas que tienen tiempos de espera o bloqueos, como la lectura y escritura en archivos, consultas a bases de datos o solicitudes a servidores. La concurrencia puede mejorar significativamente el rendimiento en estos casos, ya que permite que otras partes del programa sigan ejecutándose mientras se espera a que una tarea termine.

#### Modelos de Concurrencia en Python

Python ofrece varios enfoques para implementar concurrencia:

1. **Hilos (Threads)**:
   - Los hilos son "subprocesos" de un proceso principal que permiten ejecutar múltiples tareas dentro de ese proceso. Python usa el módulo `threading` para implementar hilos.
   - Un problema en Python es el **Global Interpreter Lock (GIL)**, que limita la ejecución de un solo hilo a la vez dentro de un proceso. Esto significa que el uso de hilos en Python no ofrece verdadera concurrencia en tareas de CPU, pero puede ser eficaz en tareas de I/O donde el programa puede hacer otras cosas mientras espera.
   
2. **Multiprocesamiento (Multiprocessing)**:
   - El módulo `multiprocessing` permite crear múltiples procesos, cada uno con su propia instancia de Python, evitando el problema del GIL.
   - Esto es útil para tareas intensivas en CPU, ya que cada proceso puede ejecutarse en paralelo en diferentes núcleos del CPU.
   - El multiprocesamiento puede requerir más memoria y recursos, ya que cada proceso es independiente, pero permite la ejecución simultánea de tareas.

3. **Asyncio (Asincronía)**:
   - El módulo `asyncio` permite escribir código asíncrono utilizando `async` y `await`, lo que facilita la programación concurrente de tareas sin bloquear la ejecución.
   - A diferencia de los hilos y procesos, `asyncio` permite la ejecución de tareas sin crear múltiples hilos o procesos. Usa un único hilo de ejecución, alternando entre tareas según su disponibilidad.
   - Es especialmente útil para aplicaciones con muchas tareas de entrada/salida, como servidores web, ya que permite que las tareas se alternen sin bloquearse entre sí.

#### Ejemplos y Casos de Uso

1. **Tareas de I/O**:
   - La concurrencia es ideal para tareas de entrada/salida, como leer y escribir archivos, acceder a bases de datos o hacer solicitudes HTTP. Las funciones asíncronas y los hilos son eficaces aquí.
   
2. **Tareas intensivas de CPU**:
   - Para el procesamiento de imágenes, cálculos matemáticos intensivos o procesamiento de grandes cantidades de datos, el multiprocesamiento es preferible, ya que puede utilizar múltiples núcleos de CPU.

3. **Programación Web y Redes**:
   - La concurrencia es muy utilizada en aplicaciones de red, como servidores que manejan múltiples clientes al mismo tiempo. `asyncio` es particularmente eficaz en estos casos para manejar múltiples conexiones sin bloquear el servidor.

#### Ventajas y Desventajas

- **Ventajas**:
  - Mejora el rendimiento en aplicaciones de I/O y en tareas que pueden realizarse en paralelo.
  - Hace un mejor uso de los recursos del sistema y reduce los tiempos de espera.

- **Desventajas**:
  - La programación concurrente es compleja y puede introducir errores difíciles de depurar, como condiciones de carrera y bloqueos.
  - Cada método tiene sus propias limitaciones: el GIL en hilos, mayor consumo de recursos en multiprocesamiento y una curva de aprendizaje en `asyncio`.

### [Hashabilidad y Diccionarios en Python](https://realpython.com/python-dicts/#hashability-and-equality)

En Python, los diccionarios son estructuras de datos clave-valor que requieren que cada clave sea **hashable**. La hashabilidad y la igualdad son esenciales para que los diccionarios puedan almacenar y buscar claves de manera eficiente. Aquí tienes una explicación detallada sobre estos conceptos:

#### 1. ¿Qué es la Hashabilidad?

Un objeto es **hashable** si su valor no cambia durante su tiempo de vida. Esto significa que:
   - Tiene un valor de **hash** constante, calculado usando la función `hash()`.
   - Se puede comparar con otros objetos a través de operadores de igualdad (`==`), y este valor de comparación debe ser consistente.

En términos prácticos:
   - Tipos de datos inmutables como `int`, `float`, `str`, y `tuple` (si contienen solo objetos inmutables) son hashables.
   - Los tipos de datos mutables, como `list`, `dict`, y `set`, no son hashables y no pueden ser usados como claves en diccionarios.

Esto se debe a que un valor hash se utiliza para determinar la ubicación de una clave en el diccionario. Si el valor de la clave cambiara, entonces también cambiaría su ubicación, y el diccionario no podría encontrarla. 

#### 2. Igualdad y Hashabilidad en los Diccionarios

Para que un objeto se pueda utilizar como clave, debe tener métodos implementados de manera consistente para:
   - `__hash__()`: Devuelve un valor hash del objeto.
   - `__eq__()`: Determina si dos objetos son iguales.

Los diccionarios en Python utilizan estos métodos para comparar claves:
   - Primero, calculan el valor hash de la clave buscada.
   - Luego, buscan ese valor hash en el diccionario. Si se encuentra un valor hash coincidente, entonces Python verifica si la clave encontrada y la clave buscada son iguales usando `__eq__()`.

**Ejemplo**:
```python
# Este es un diccionario en Python
my_dict = {1: "uno", "two": "dos", (3, 4): "tres y cuatro"}
```
- Las claves `1`, `"two"` y `(3, 4)` son hashables, ya que son inmutables y tienen un valor hash constante.

#### 3. Implicaciones de Hashabilidad para Objetos Personalizados

Si creas una clase personalizada y quieres usar instancias de esa clase como claves de diccionario, debes implementar `__hash__()` y `__eq__()` de manera adecuada:
   - **`__hash__()`** debería devolver un valor que sea único para cada instancia (por ejemplo, utilizando un atributo inmutable).
   - **`__eq__()`** debe definir claramente cuándo dos instancias son iguales.

Ejemplo de una clase con `__hash__` y `__eq__` implementados:
```python
class MyClass:
    def __init__(self, value):
        self.value = value
    
    def __hash__(self):
        return hash(self.value)
    
    def __eq__(self, other):
        return isinstance(other, MyClass) and self.value == other.value

# Ahora podemos usar instancias de MyClass como claves en un diccionario
my_obj1 = MyClass(10)
my_obj2 = MyClass(10)

my_dict = {my_obj1: "value1"}
print(my_dict[my_obj2])  # Esto imprimirá "value1", ya que my_obj1 y my_obj2 son iguales
```

#### Resumen

- La **hashabilidad** permite que las claves de diccionarios tengan un valor constante, lo que facilita la búsqueda rápida.
- La **igualdad** garantiza que Python pueda verificar si dos claves representan el mismo valor, incluso si son instancias diferentes.
- Las claves en los diccionarios de Python deben ser inmutables y hashables para asegurar que el diccionario funcione de manera correcta y eficiente.
