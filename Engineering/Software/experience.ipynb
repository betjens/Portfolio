## 1. Algorithms & Data Structures

### Basic
Fundamental building blocks for solving computational problems efficiently.

- **Basic structures:** Arrays, Linked Lists, Stacks, Queues, Hash Tables, Trees, Graphs.
  - *Arrays* store elements in contiguous memory for O(1) index access. 
  - *Linked Lists* use pointers for efficient insertions/deletions. 
  - *Stacks* (LIFO) and *Queues* (FIFO) control element ordering. 
  - *Hash Tables* offer O(1) average lookups via key hashing. 
  - *Trees* model hierarchical data (BSTs enable O(log n) search). 
  - *Graphs* represent networks of nodes and edges for relationship modeling.

- **Basic algorithms:** Linear/Binary Search, Bubble/Merge/Quick Sort.
  - *Linear Search* scans every element — O(n), no prerequisites. 
  - *Binary Search* halves the search space each step — O(log n), requires sorted input. 
  - *Bubble Sort* repeatedly swaps adjacent elements — simple but O(n²). 
  - *Merge Sort* divides and conquers — stable and O(n log n). 
  - *Quick Sort* partitions around a pivot — O(n log n) average, often fastest in practice.
```python
# Binary Search example
def binary_search(arr, target):
    lo, hi = 0, len(arr) - 1
    while lo <= hi:
        mid = (lo + hi) // 2
        if arr[mid] == target: return mid
        elif arr[mid] < target: lo = mid + 1
        else: hi = mid - 1
    return -1
```

### Advanced
Knowledge of implementation details: how a Hash Table handles collisions (chaining vs open addressing), tree balancing (AVL, Red-Black), graph traversal (BFS/DFS), heaps, tries, segment trees.

- **Hash Table collisions:** *Chaining* stores colliding keys in a linked list per bucket — simple but memory-heavy. *Open Addressing* probes for the next free slot (linear, quadratic, or double hashing) — cache-friendly but degrades at high load factors.
- **Tree balancing:** *AVL trees* enforce a strict height difference of ≤1 between subtrees via rotations — faster lookups. *Red-Black trees* use color rules for a looser balance — fewer rotations, faster insertions/deletions (used in most stdlib maps).
- **Graph traversal:** *BFS* explores level by level using a queue — ideal for shortest path in unweighted graphs. *DFS* goes deep via a stack/recursion — useful for cycle detection, topological sort, and connected components.
- **Heaps:** A complete binary tree satisfying the heap property — *Min-Heap* for priority queues and Dijkstra's algorithm, *Max-Heap* for scheduling. Backed by an array for cache efficiency.
- **Tries:** A prefix tree where each node represents a character — enables O(m) search/insert by word length m, regardless of dictionary size. Used in autocomplete and spell checkers.
- **Segment Trees:** A binary tree over an array that stores aggregate values (sum, min, max) per range — supports both range queries and point updates in O(log n). Essential for interval problems.

**Example — Min-Heap property:** every parent node ≤ its children, enabling O(log n) extraction of the minimum.

## 2. Network Concepts

### Fundamentals
OSI model layers, TCP/IP stack, DNS resolution, HTTP/HTTPS, ports, sockets, routing basics.

- **OSI model layers:** A 7-layer framework (Physical → Data Link → Network → Transport → Session → Presentation → Application) that standardizes how data travels between systems — each layer serves the one above it and is served by the one below.
- **TCP/IP stack:** The practical 4-layer implementation of OSI (Network Access, Internet, Transport, Application) that powers the modern internet — TCP guarantees ordered, reliable delivery; UDP trades reliability for speed.
- **DNS resolution:** Translates human-readable hostnames (e.g. `google.com`) into IP addresses via a hierarchy of resolvers, root servers, and authoritative nameservers — results are cached at each step to reduce latency.
- **HTTP/HTTPS:** HTTP is a stateless request/response protocol for transferring hypertext; HTTPS wraps it in TLS for encrypted, authenticated communication — mandatory for any data that must stay private in transit.
- **Ports:** A 16-bit number (0–65535) that identifies a specific process on a host — well-known ports include 80 (HTTP), 443 (HTTPS), 22 (SSH), and 53 (DNS).
- **Sockets:** An endpoint defined by an IP address + port pair — the OS abstraction that lets processes send and receive data over a network as if reading/writing a file.
- **Routing basics:** Routers forward packets hop-by-hop using routing tables and protocols (e.g. BGP, OSPF) to find the best path across networks — decisions are based on destination IP, not the full route.

| Layer | Example Protocol | Role |
|-------|-----------------|------|
| Application | HTTP, DNS, FTP | User-facing data exchange |
| Transport | TCP, UDP | End-to-end delivery & flow control |
| Network | IP, ICMP | Addressing & packet routing |
| Data Link | Ethernet, Wi-Fi | Node-to-node transfer on the same network |

### Network Principles & Configuration
- Subnetting, VLANs, NAT, firewalls, load balancers.
  - *Subnetting* divides an IP space into smaller networks using CIDR masks (e.g. `/24`) — limits broadcast domains and improves security by isolating segments. 
  - *VLANs* achieve logical separation on the same physical switch — traffic between VLANs requires a router or Layer 3 switch.
  - *NAT* maps private IPs to a public one, preserving the IPv4 address space and acting as an implicit firewall. 
  - *Firewalls* filter traffic by rules (IP, port, protocol, state) — stateful firewalls also track active connections. 
  - *Load balancers* distribute incoming requests across backend instances using strategies like round-robin, least-connections, or IP hash — improving throughput and eliminating single points of failure.

- Ability to configure routers/switches, set up VPNs, troubleshoot connectivity with tools like `ping`, `traceroute`, `netstat`, `tcpdump`.
  - *Router/switch configuration* covers assigning interfaces, setting static or dynamic routes (OSPF, BGP), and managing ACLs.
  - *VPNs* tunnel encrypted traffic over public networks — site-to-site VPNs link entire networks while remote-access VPNs (WireGuard, OpenVPN) serve individual clients.
  - *`ping`* tests reachability via ICMP echo.
  - *`traceroute`* maps the hop-by-hop path to a destination and surfaces where latency spikes.
  - *`netstat`* (or `ss`) lists open sockets, active connections, and listening ports. 
  - *`tcpdump`* captures raw packets on an interface — the go-to tool for inspecting actual traffic and diagnosing protocol-level issues.

## 3. Cryptography

### Importance
Cryptography protects confidentiality, integrity, and authenticity of data in transit and at rest. It underpins HTTPS, authentication tokens, password storage, and digital signatures.

### Symmetric-Key Encryption
Same key used to encrypt and decrypt. Fast, suitable for bulk data.

- **Examples:** AES-256, ChaCha20.
- **Use case:** Encrypting files or database columns.

```
plaintext + key → [AES encrypt] → ciphertext
ciphertext + key → [AES decrypt] → plaintext
```

### Public-Key (Asymmetric) Encryption
A key pair: public key encrypts, private key decrypts (or vice versa for signatures).

- **Examples:** RSA, ECDSA, Diffie-Hellman key exchange.
- **Use case:** TLS handshake — server shares public key, client uses it to exchange a session key, then symmetric encryption takes over.

## 4. Databases

### Core Concepts
Tables (relations), ACID properties, indexes, normalization (1NF–3NF), transactions, foreign keys.

- **Tables (relations):** The fundamental unit of a relational database — data is organized into rows (records) and columns (attributes), with each table representing a single entity type. Relationships between tables are expressed via shared keys.
- **ACID properties:** *Atomicity* ensures a transaction fully succeeds or fully rolls back. *Consistency* guarantees the database moves from one valid state to another. *Isolation* prevents concurrent transactions from interfering with each other. *Durability* ensures committed changes survive crashes — typically enforced via write-ahead logging (WAL).
- **Indexes:** Auxiliary data structures (commonly B-trees) that allow the engine to locate rows without a full table scan — dramatically speeds up reads at the cost of slower writes and extra storage. Over-indexing is a common performance pitfall.
- **Normalization (1NF–3NF):** A process of structuring tables to eliminate redundancy. *1NF* requires atomic column values and unique rows. *2NF* removes partial dependencies on composite keys. *3NF* eliminates transitive dependencies — every non-key column depends only on the primary key.
- **Transactions:** A sequence of operations executed as a single unit — either all committed or all rolled back. Managed via `BEGIN`, `COMMIT`, and `ROLLBACK` statements.
- **Foreign keys:** A column referencing the primary key of another table — enforces referential integrity so no orphaned records can exist. Can cascade updates/deletes automatically.

## 5. Virtualization & Containerization

### Virtualization
A **hypervisor** (VMware, VirtualBox, Hyper-V, KVM) abstracts physical hardware to run multiple isolated **Virtual Machines (VMs)**, each with its own OS kernel.

**Key operations:** create/configure VMs, take **snapshots** (point-in-time state), set up **backups**, attach virtual devices (disks, NICs).

| Type | Description | Example |
|------|-------------|---------|
| Type 1 (Bare-metal) | Runs directly on hardware | VMware ESXi, KVM |
| Type 2 (Hosted) | Runs on host OS | VirtualBox, VMware Workstation |

### Containerization
Containers share the host OS kernel but isolate processes via namespaces and cgroups. Lighter and faster to start than VMs.

```bash
# Build and run a Docker container
docker build -t myapp:1.0 .
docker run -d -p 8080:80 --name myapp myapp:1.0

# Docker Compose for multi-container apps
docker-compose up -d
```

| | VMs | Containers |
|--|-----|------------|
| Isolation | Full OS | Process-level |
| Startup | Minutes | Seconds |
| Size | GBs | MBs |
| Overhead | Higher | Lower |

### Deployment Strategies

| Strategy | Description | Pros / Cons |
|----------|-------------|-------------|
| **Rolling** | Gradually replace old instances | No downtime; rollback is slow |
| **Blue/Green** | Two identical envs, switch traffic | Instant rollback; doubles infra cost |
| **Canary** | Route small % of traffic to new version | Low risk; complex routing |
| **Recreate** | Stop all, then deploy new | Simple; causes downtime |

**Selection guidance:** Blue/Green for critical zero-downtime services; Canary for high-risk feature releases; Rolling for general microservices; Recreate for dev/staging.

- **Memory leaks:** heap profilers (VisualVM, Valgrind).
- **Performance:** CPU profilers (`perf`, py-spy, async-profiler).

## 7. Application Profiling

Profiling identifies bottlenecks in CPU, memory, I/O, or network. Always profile before optimizing — gut instinct about where the slowdown lives is wrong more often than not.

- **CPU profiling:** flame graphs, sampling vs instrumentation.
  - *Flame graphs* visualize the full call stack over time — the width of each bar represents how often that function appeared in samples, making hot paths immediately visible. *Sampling* profilers interrupt the program at regular intervals to record the current stack — low overhead, safe for production. *Instrumentation* profilers inject timing code into every function call — far more precise but can slow the program enough to skew results.

- **Memory profiling:** heap snapshots, allocation tracking.
  - *Heap snapshots* capture the full object graph at a point in time — compare two snapshots before and after a suspected leak to find objects that should have been garbage collected but weren't. *Allocation tracking* records every allocation with its call stack over a time window — useful for finding which code paths are putting the most pressure on the GC, causing excessive pauses or churn.

- **Tools:** JProfiler, YourKit (Java), Py-Spy (Python), Chrome DevTools (frontend), `perf` (Linux system).
  - *JProfiler* and *YourKit* offer deep JVM integration — CPU, memory, threads, and DB call tracing in one UI. *Py-Spy* is a sampling profiler for Python that attaches to a running process without restarting it — ideal for diagnosing live production slowdowns. *Chrome DevTools* covers frontend CPU (Performance tab), memory (Memory tab), and network waterfalls in the browser. *`perf`* is a Linux kernel-level tool for hardware counter sampling, cache misses, and syscall tracing — the lowest-level view available without a dedicated APM.

**Workflow:** run profiler → identify hot path → optimize (algorithm, caching, DB query) → verify improvement.
  - Never skip the final *verify* step — optimizations sometimes shift the bottleneck elsewhere rather than eliminating it. Re-run the profiler under the same load conditions to confirm the target metric (latency, throughput, memory) actually improved, and keep the baseline numbers for future comparison.

## 8. IDE Usage

### Core Features to Master
- **Refactoring tools:** rename, extract method/variable, inline.
  - *Rename* updates every reference across the codebase in one step — safer than find-and-replace which ignores scope. *Extract method/variable* isolates repeated or complex expressions into named, reusable units — the fastest way to improve readability. *Inline* does the reverse, collapsing a trivial variable or method back into its call site to reduce indirection.

- **Debugger:** breakpoints, watch expressions, step over/into/out.
  - *Breakpoints* pause execution at a specific line — conditional breakpoints (e.g. break only when `i == 500`) are especially powerful for hard-to-reproduce bugs. *Watch expressions* evaluate arbitrary expressions live as you step through code. *Step over* executes the current line without entering calls; *step into* dives inside a function call; *step out* runs to the end of the current function and returns to the caller.

- **Code navigation:** go-to definition, find usages, call hierarchy.
  - *Go-to definition* jumps instantly to where a symbol is declared — essential for exploring unfamiliar codebases. *Find usages* lists every reference to a symbol across the project, making impact analysis before refactoring safe and fast. *Call hierarchy* visualizes who calls a function and who those callers are called by — invaluable for tracing execution paths.

- **Version control integration:** diff view, blame, resolve conflicts.
  - *Diff view* shows line-by-line changes between commits or branches directly in the editor — faster than switching to a terminal. *Blame* (or *annotate*) reveals who last modified each line and in which commit — pairs well with the commit message for context. *Conflict resolution* lets you pick incoming, current, or a manual merge per conflict hunk without leaving the IDE.

- **Code generation:** live templates, postfix completions.
  - *Live templates* expand short abbreviations into boilerplate snippets (e.g. `fori` → a full for-loop) — define custom ones for patterns repeated in your codebase. *Postfix completions* transform an already-typed expression (e.g. `items.for`) into a surrounding construct — keeping your cursor flow forward and reducing backtracking.

- **Static analysis & linting:** inspections, quick-fix suggestions.
  - *Inspections* run continuously in the background, flagging bugs, code smells, and style violations without a separate tool invocation. *Quick-fix suggestions* (the lightbulb / `Alt+Enter`) offer one-click resolutions — from adding a missing import to inverting a condition — and are often faster than typing the fix manually.

## 9. Software Licenses

### Main License Types

| License | Copyleft | Commercial Use | Key Requirement |
|---------|----------|----------------|-----------------|
| MIT | No | ✅ | Keep copyright notice |
| Apache 2.0 | No | ✅ | Notice + patent grant |
| GPL v2/v3 | Strong | ✅ (with conditions) | Derivative works must be GPL |
| LGPL | Weak | ✅ | Linking allowed; modifications to lib must be open |
| AGPL | Network copyleft | ✅ (with conditions) | SaaS use triggers open-source requirement |
| BSD 2/3-Clause | No | ✅ | Keep copyright notice |
| Proprietary | N/A | Defined by owner | Must comply with EULA |

### Decision Framework
Before using a dependency, ask:
1. Is the license OSI-approved or proprietary?
2. Does it have copyleft? Would it force our code to be open-sourced?
3. Is the intended use (SaaS, internal tool, redistributed product) compatible?
4. Are there patent clauses that could affect the company?

> **Example:** Using a GPL library in a closed-source commercial product is generally incompatible unless a commercial license is obtained. LGPL is usually safe for dynamic linking in proprietary software.
