# Software Engineering Processes

---

## 1. SDLC — Software Development Life Cycle

The SDLC defines the structured stages a software product goes through from inception to retirement.

**Major Stages:**
1. **Planning** — define scope, feasibility, resources
2. **Requirements** — gather and document what needs to be built
3. **Design** — architecture, UI/UX, database design
4. **Implementation** — actual coding
5. **Testing** — QA, bug fixing, validation
6. **Deployment** — release to production
7. **Maintenance** — monitoring, patches, enhancements

**Key Roles:** Product Owner, Business Analyst, Architect, Developer, QA Engineer, DevOps, Project Manager, Scrum Master.

---

## 2. Secure SDLC (S-SDLC)

S-SDLC embeds security practices into every phase of the SDLC rather than treating security as an afterthought.

**Purpose:** Reduce vulnerabilities, minimize breach risk, ensure compliance (e.g., OWASP, GDPR).

**Examples of practices per phase:**
- *Requirements* → define security requirements (auth, encryption)
- *Design* → threat modeling (e.g., STRIDE)
- *Implementation* → secure coding guidelines, code linters
- *Testing* → SAST, DAST, penetration testing
- *Deployment* → secrets management, secure CI/CD pipelines

## 3. Development Methodologies

### Waterfall
Linear, sequential approach. Each phase must be completed before the next begins.

- ✅ Clear structure, good for fixed-scope projects
- ❌ Inflexible to change; issues found late are costly

### Agile
Iterative, incremental delivery. Work is broken into short cycles (sprints/iterations).

- ✅ Adapts to change, continuous feedback, faster delivery
- ❌ Requires close collaboration; harder to predict long-term timelines

---

## 4. Agile Frameworks

### Scrum
Time-boxed sprints (1–4 weeks) with defined roles and ceremonies.

| Roles | Events | Artifacts |
|---|---|---|
| Product Owner | Sprint Planning | Product Backlog |
| Scrum Master | Daily Standup | Sprint Backlog |
| Dev Team | Sprint Review | Increment |
| | Retrospective | |

- ✅ Predictable cadence, clear accountability
- ❌ Can become rigid if team doesn't adapt; requires committed PO

### Kanban
Visual board-based flow with no fixed iterations. Work items pulled as capacity allows.

- ✅ Flexible, great for support/maintenance work
- ❌ Less structured; can lack urgency or sprint goals

## 5. Requirements

### Functional Requirements
Define *what* the system does.
> *"A user must be able to reset their password via email."*

### Non-Functional Requirements (NFRs)
Define *how well* the system performs — quality attributes.
> *"The login page must load in under 2 seconds for 95% of requests."*

Common NFR categories: Performance, Security, Scalability, Reliability, Usability, Maintainability.

**Identifying & controlling NFRs:**
- Elicit them during requirements workshops
- Make them measurable (avoid vague terms like "fast" or "secure")
- Include NFR validation in acceptance criteria and Definition of Done

### Acceptance Criteria
Define the conditions a feature must meet to be accepted by the stakeholder.

> **User Story:** As a user, I want to log in with my email.
>
> **Acceptance Criteria:**
> - Given valid credentials → user is redirected to dashboard
> - Given invalid credentials → error message is shown
> - Given 5 failed attempts → account is locked for 15 minutes

## 6. Definition of Ready (DoR) & Definition of Done (DoD)

| | DoR | DoD |
|---|---|---|
| **Purpose** | Ensures a task is ready to be worked on | Ensures a task is truly complete |
| **Example** | Story is written, AC defined, designs available | Code reviewed, tests passing, deployed to staging, docs updated |

These prevent ambiguity and maintain quality gates across the team.

## 7. Issue Tracking Systems

Tools to manage tasks, bugs, and workflows.

- **JIRA** — most popular for Agile teams; supports Scrum boards, Kanban, epics, sprints
- **Linear** — fast, developer-friendly UI
- **GitHub Issues / GitLab** — integrated with code repositories
- **Azure DevOps** — full ALM suite, common in enterprise .NET environments
- **Trello** — simple Kanban boards, good for small teams

## 8. Estimation Techniques

### Estimation Notions
- **Absolute** — hours, days (e.g., "3 days")
- **Relative** — story points, t-shirt sizes (e.g., "M vs L", "5 vs 8 points")

### Techniques Overview

| Technique | Best For | Example |
|---|---|---|
| **By Analogy** | Similar past work | "This feature is like the login we built — 3 days" |
| **Expert Judgment** | Complex unknowns | Senior dev gives estimate based on experience |
| **Planning Poker** | Team alignment | Each member picks a card; discuss outliers |
| **Decomposition** | Large tasks | Break epic into stories, estimate each |
| **Bucket System** | Many items quickly | Sort stories into size buckets collaboratively |
| **T-Shirt Sizing** | Early rough estimates | XS / S / M / L / XL |
| **Story Points** | Relative complexity | Fibonacci sequence: 1, 2, 3, 5, 8, 13… |

### Estimating Unknown Tasks
- Time-box a spike (research task) to reduce uncertainty
- Use range estimates: "3–7 days"
- Apply risk buffer: add 20–30% for unknowns
- Reassess after the first iteration

### Handling Under/Over Estimation
- **Under-estimated:** Break down further, add spikes, negotiate scope
- **Over-estimated:** Analyze why, recalibrate reference stories, remove padding
- Track **velocity** over sprints to improve future accuracy

## 9. Work Planning & Decomposition

For significant scope (component/module/layer):

1. Identify the high-level goal and boundaries
2. Break into epics → user stories → tasks
3. Estimate at story level using relative sizing
4. Map dependencies and risks
5. Prioritize by business value + technical dependencies

> **Example:** Building a "User Profile" module
> - Epic: User Profile Management
>   - Story: View profile (S)
>   - Story: Edit profile (M)
>   - Story: Upload avatar (M)
>   - Story: Delete account (L)

## 10. Key Project Metrics & Delivery Predictability

| Metric | What It Tells You |
|---|---|
| **Velocity** | Average story points completed per sprint |
| **Cycle Time** | Time from task start to completion |
| **Lead Time** | Time from request to delivery |
| **Burndown Chart** | Sprint progress toward goal |
| **Throughput** | Number of items completed per period |
| **Defect Rate** | Quality signal — bugs per release |

**How to use metrics:**
- Stable velocity → reliable sprint forecasting
- Rising cycle time → identify blockers or scope creep
- High defect rate → invest in testing or DoD tightening
- Don't use metrics to punish; use them to improve the process

## 11. Process Improvement

**Retrospectives** are the primary Agile mechanism — team reflects on what went well, what didn't, and what to try next.

**Suggesting improvements:**
- Base suggestions on observed data (metrics, team feedback)
- Use frameworks like **Start / Stop / Continue** or **4Ls** (Liked, Learned, Lacked, Longed for)
- Propose one or two small experiments per retro, not a full process overhaul
- Measure impact of changes in the next sprint

**Establishing a process from scratch:**
1. Understand team size, project type, stakeholder availability
2. Choose a base framework (Scrum / Kanban)
3. Define DoR, DoD, and ceremonies
4. Set up tooling (JIRA, Git branching strategy, CI/CD)
5. Run for 2–3 sprints, then retrospect and adapt

## 12. Secure Development Practices

Key practices to embed security in the team's daily work:

- **Threat Modeling** — identify risks during design (e.g., STRIDE model)
- **Secure Coding Standards** — OWASP Top 10 awareness, input validation, least privilege
- **Code Review with Security Lens** — check for injection, auth flaws, sensitive data exposure
- **Dependency Scanning** — tools like Snyk, Dependabot
- **Secrets Management** — no hardcoded credentials; use vaults (HashiCorp Vault, AWS Secrets Manager)
- **Security in CI/CD** — integrate SAST (e.g., SonarQube) and DAST tools
- **Security Training** — keep team updated on common vulnerabilities

**Promoting adoption:**
- Add security checks to the DoD
- Include security stories in backlog refinement
- Make security a team responsibility, not just a security team's job
